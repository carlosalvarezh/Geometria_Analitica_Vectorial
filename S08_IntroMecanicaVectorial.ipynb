### Definición y Campo de Estudio de la Mecánica


La mecánica es una ciencia que estudia los cuerpos en reposo o en movimiento bajo la acción de fuerzas. Es esencial en el campo de la ingeniería, pues proporciona los métodos para analizar, diseñar y optimizar estructuras y dispositivos mecánicos. Se divide en tres ramas principales:

- ***Mecánica de Cuerpos Rígidos:*** Comprende la estática y la dinámica. La estática se enfoca en cuerpos en reposo, analizando condiciones de equilibrio. La dinámica estudia los cuerpos en movimiento, considerando las leyes de Newton y otras teorías relacionadas.

### La Mecánica como Ciencia Física y Aplicada


Aunque algunos consideran a la mecánica como parte de las matemáticas debido a su énfasis en el razonamiento deductivo y el uso de modelos matemáticos, en ingeniería, se aprecia como una ciencia aplicada. Su objetivo es no solo entender los principios teóricos sino también aplicarlos en soluciones prácticas y efectivas en ingeniería.

### Aplicaciones de la Mecánica en Ingeniería Civil

- **Estática:** Fundamental para el diseño de puentes, represas y otros elementos estructurales, donde se debe asegurar que puedan soportar las cargas aplicadas sin fallar.
<p>&nbsp;</p>

- **Dinámica:** Esencial para el diseño de estructuras que deben resistir movimientos, como edificios en zonas sísmicas, así como para el análisis de la respuesta de estructuras y máquinas a cargas dinámicas.



### Conceptos y Principios Fundamentales de la Mecánica

#### Evolución Histórica de los Principios de la Mecánica

El desarrollo de la mecánica como ciencia formal comenzó con pensadores como Aristóteles y Arquímedes, pero no fue hasta Isaac Newton en el siglo XVII que los principios fundamentales de la mecánica fueron formalizados. Estos principios, posteriormente modificados por d'Alembert, Lagrange y Hamilton, han sido la base de las ciencias de la ingeniería hasta las adaptaciones requeridas por la teoría de la relatividad de Einstein en 1905. La mecánica newtoniana sigue siendo crucial en la enseñanza de la ingeniería, a pesar de sus limitaciones en escenarios extremos.

#### Conceptos Básicos de la Mecánica

La mecánica se apoya en cuatro conceptos fundamentales:
- **Espacio:** Se refiere a la posición de un punto \(P\) definido por sus coordenadas en relación a un origen.
<p>&nbsp;</p>

- **Tiempo:** Necesario para especificar la temporalidad de los eventos.
<p>&nbsp;</p>

- **Masa:** Es una medida que caracteriza la cantidad de materia y afecta cómo los cuerpos experimentan la fuerza gravitacional y la inercia.
<p>&nbsp;</p>

- **Fuerza:** Una fuerza representa la acción de un cuerpo sobre otro y puede ejercerse por contacto real como empujar o jalar, a distancia como en el caso de las fuerzas gravitacionales y magnéticas. Una fuerza se caracteriza por su punto de aplicación, magnitud y dirección, y se representa con un vector que indica la interacción entre cuerpos.


Se estudiarán las condiciones de reposo o movimiento de partículas y cuerpos rígidos a partir de los cuatro conceptos básicos que se han expuesto. Por **partícula** se entiende una pequeñísima cantidad de materia que ocupa un punto en el espacio. Un **cuerpo rígido** es la combinación de un gran número de partículas que ocupan posiciones fijas entre sí. El estudio de la mecánica de las partículas es un requisito previo al de los cuerpos rígidos. Además, los resultados obtenidos para una partícula pueden usarse directamente en muchos problemas que tratan de las condiciones de reposo o movimiento de cuerpos reales.

#### Principios de la Mecánica Newtoniana

El estudio de la mecánica elemental descansa sobre seis principios fundamentales basados en la evidencia experimental.

- **La ley del paralelogramo para la adición de fuerzas:** Establece que dos fuerzas que actúan sobre una partícula pueden ser sustituidas por una sola fuerza llamada resultante, que se obtiene al trazar la diagonal del paralelogramo que tiene los lados iguales a las fuerzas dadas.
<p>&nbsp;</p>

- **El principio de transmisibilidad:** Establece que las condiciones de equilibrio o de movimiento de un cuerpo rígido permanecerán inalteradas si una fuerza que actúa en un punto del cuerpo rígido se sustituye por una fuerza de la misma magnitud y la misma dirección, pero que actúe en un punto diferente, siempre que las dos fuerzas tengan la misma línea de acción.
<p>&nbsp;</p>

- **Primera Ley de Newton (Ley de Inercia):** 
     $$\text{Si } \sum \mathbf{F} = 0, \text{ entonces } \mathbf{v} = \text{constante}$$
<p>&nbsp;</p>

- **Segunda Ley de Newton (Ley de Fuerza):** Describe cómo la aceleración de un cuerpo es proporcional a la fuerza neta aplicada y ocurre en la dirección de la fuerza.
     $$\mathbf{F} = m \mathbf{a}$$
<p>&nbsp;</p>
     
- **Tercera Ley de Newton (Acción y Reacción):** Cada acción tiene una reacción igual y opuesta.
     $$\mathbf{F}_{\text{acción}} = -\mathbf{F}_{\text{reacción}}$$
<p>&nbsp;</p>
     
- **Ley de Gravitación Universal:** La fuerza entre dos masas es directamente proporcional al producto de las masas e inversamente proporcional al cuadrado de la distancia entre ellas.
     $$F = G \frac{M m}{r^2}$$
     Donde $G$ es la constante de gravitación universal.



### Sistema de unidades

#### Introducción

Los cuatro conceptos fundamentales introducidos en la sección anterior se asocian las llamadas unidades cinéticas, es decir, las unidades de *longitud*, *tiempo*, *masa* y *fuerza*. Estas unidades no pueden escogerse de manera independiente si la ecuación dada por la segunda ley de Newton ($\mathbf{F} = m \mathbf{a}$) ha de satisfacerse. Tres de ellas pueden definirse en forma arbitraria; se les llama unidades básicas. La cuarta unidad, en cambio, debe escogerse de acuerdo con la ecuación $\mathbf{F} = m \mathbf{a}$ y se le identifica como unidad derivada. Se dice que las unidades cinéticas así seleccionadas forman un sistema consistente de unidades.

#### Sistema Internacional de Unidades (SI)

**Unidades Básicas del SI:**
El Sistema Internacional de Unidades (SI) es un sistema universal utilizado en todo el mundo. Las unidades básicas del SI para longitud, masa y tiempo son:
- **Metro (m):** Originalmente definido como la diezmillonésima parte de la distancia del ecuador al polo, ahora se define como $1,650,763.73$ longitudes de onda de la luz naranja-roja de la transición en el átomo de *criptón-86*.
- **Kilogramo (kg):** Aproximadamente igual a la masa de $0.001 m^3$ de agua, se define como la masa de un patrón de platino-iridio guardado en Sèvres, Francia.
- **Segundo (s):** Inicialmente definido como $1/86,400$ del día solar medio, ahora se define como la duración de $9,192,631,770$ ciclos de la radiación emitida por la transición entre dos niveles del estado fundamental del átomo de *cesio-133*.

**Unidad de Fuerza:**
La unidad derivada de fuerza en el SI es el newton (N). Se define como la fuerza necesaria para proporcionar una aceleración de 1 m/s² a una masa de 1 kg:
$$
1 \, \text{N} = (1 \, \text{kg}) \left(1 \, \text{m/s}^2\right) = 1 \, \text{kg} \cdot \text{m/s}^2
$$

**Sistema Absoluto de Unidades:**
El SI es un sistema absoluto porque las unidades de metro, kilogramo y segundo son independientes del lugar de medición. Estas unidades mantienen su significado en cualquier lugar de la Tierra e incluso en otros planetas.

**Peso en el SI:**
El peso, o la fuerza de gravedad ejercida sobre un cuerpo, se expresa en newtons. Para un cuerpo de masa 1 kg, el peso se calcula usando la aceleración debida a la gravedad ($g$):
$$
\begin{aligned}
W & = m g \\
& = (1 \, \text{kg}) \left(9.81 \, \text{m/s}^2\right) \\
& = 9.81 \, \text{N}
\end{aligned}
$$

**Prefijos y Unidades Derivadas:**
Los múltiplos y submúltiplos de las unidades fundamentales del SI se obtienen usando prefijos. Los más comunes en ingeniería son:
- **Kilómetro (km):** 
  $$
  1 \, \text{km} = 1000 \, \text{m}
  $$
- **Milímetro (mm):**
  $$
  1 \, \text{mm} = 0.001 \, \text{m}
  $$
- **Megagramo (Mg):**
  $$
  1 \, \text{Mg} = 1000 \, \text{kg}
  $$
- **Gramo (g):**
  $$
  1 \, \text{g} = 0.001 \, \text{kg}
  $$
- **Kilonewton (kN):**
  $$
  1 \, \text{kN} = 1000 \, \text{N}
  $$

La conversión entre estas unidades y las unidades base (metros, kilogramos, newtons) se realiza moviendo el punto decimal tres lugares:
- **De kilómetros a metros:**
  $$
  3.82 \, \text{km} = 3820 \, \text{m}
  $$
- **De milímetros a metros:**
  $$
  47.2 \, \text{mm} = 0.0472 \, \text{m}
  $$

En notación científica:
$$
\begin{aligned}
3.82 \, \text{km} & = 3.82 \times 10^3 \, \text{m} \\
47.2 \, \text{mm} & = 47.2 \times 10^{-3} \, \text{m}
\end{aligned}
$$

**Múltiplos de Tiempo:**
Los múltiplos de la unidad de tiempo son el minuto (min) y la hora (h):
- **Minuto:**
  $$
  1 \, \text{min} = 60 \, \text{s}
  $$
- **Hora:**
  $$
  1 \, \text{h} = 60 \, \text{min} = 3600 \, \text{s}
  $$

**Uso de Prefijos Adecuados:**
Utilizar los prefijos adecuados ayuda a evitar números muy grandes o muy pequeños:
- **$427.2 \text{ km}$** en lugar de **$427200 \text{ m}$**.
- **$2.16 \text{ mm}$** en lugar de **$0.00216 \text{ m}$**.

Este sistema de unidades facilita la conversión y estandarización de medidas en diversos campos científicos y de ingeniería, promoviendo una comprensión y comunicación claras y precisas.

#### Unidades de área y volumen

La unidad de área es el metro cuadrado $\left(\mathrm{m}^2\right)$, que representa el área de un cuadrado de $1 \mathrm{~m}$ de lado; la unidad de volumen es el metro cúbico $\left(\mathrm{m}^3\right)$, que es igual al volumen de un cubo de $1 \mathrm{~m}$ de lado. Para evitar valores numéricos excesivamente pequeños o demasiado grandes en el cálculo de áreas y volúmenes, se usan sistemas de subunidades que se obtienen elevando, respectivamente, al cuadrado y al cubo no sólo el milímetro sino también dos submúltiplos intermedios del metro, llamados decímetro (dm) y centímetro $(\mathrm{cm})$. Entonces, por definición,
$$
\begin{aligned}
1 \mathrm{dm} & =0.1 \mathrm{~m}=10^{-1} \mathrm{~m} \\
1 \mathrm{~cm} & =0.01 \mathrm{~m}=10^{-2} \mathrm{~m} \\
1 \mathrm{~mm} & =0.001 \mathrm{~m}=10^{-3} \mathrm{~m}
\end{aligned}
$$
los submúltiplos de la unidad de área son
$$
\begin{aligned}
1 \mathrm{dm}^2 & =(1 \mathrm{dm})^2=\left(10^{-1} \mathrm{~m}\right)^2=10^{-2} \mathrm{~m}^2 \\
1 \mathrm{~cm}^2 & =(1 \mathrm{~cm})^2=\left(10^{-2} \mathrm{~m}\right)^2=10^{-4} \mathrm{~m}^2 \\
1 \mathrm{~mm}^2 & =(1 \mathrm{~mm})^2=\left(10^{-3} \mathrm{~m}\right)^2=10^{-6} \mathrm{~m}^2
\end{aligned}
$$
y los submúltiplos de la unidad de volumen son
$$
\begin{aligned}
1 \mathrm{dm}^3 & =(1 \mathrm{dm})^3=\left(10^{-1} \mathrm{~m}\right)^3=10^{-3} \mathrm{~m}^3 \\
1 \mathrm{~cm}^3 & =(1 \mathrm{~cm})^3=\left(10^{-2} \mathrm{~m}\right)^3=10^{-6} \mathrm{~m}^3 \\
1 \mathrm{~mm}^3 & =(1 \mathrm{~mm})^3=\left(10^{-3} \mathrm{~m}\right)^3=10^{-9} \mathrm{~m}^3
\end{aligned}
$$

Debe notarse que cuando se mide el volumen de un líquido, el decímetro cúbico $\left(\mathrm{dm}^3\right)$ se conoce como un litro $(\mathrm{L})$.

En la tabla 1.2 se muestran otras unidades derivadas del SI, que se usan para medir el momento de una fuerza, el trabajo de una fuerza, etc. Aunque estas unidades se introducirán en capítulos posteriores conforme se vayan necesitando, es necesario describir una regla importante en esta fase: cuando se obtiene una unidad derivada con la división de una unidad básica entre otra unidad básica, debe usarse un prefijo en el numerador de la unidad derivada, pero no en su denominador. Por ejemplo, la constante $k$ de un resorte que se elonga $20 \mathrm{~mm}$ bajo una carga de $100 \mathrm{~N}$ se expresará como
$$
k=\frac{100 \mathrm{~N}}{20 \mathrm{~mm}}=\frac{100 \mathrm{~N}}{0.020 \mathrm{~m}}=5000 \mathrm{~N} / \mathrm{m} \quad \text { o } \quad k=5 \mathrm{kN} / \mathrm{m}
$$
pero nunca como $k=5 \mathrm{~N} / \mathrm{mm}$.

#### Unidades de uso común en Estados Unidos (USA)

***Unidades Básicas:***
- **Longitud:** Pie (ft)
- **Fuerza:** Libra (lb)
- **Tiempo:** Segundo (s) (idéntico al del SI)

***Definiciones:***
- **Pie (ft):** 0.3048 metros (m)
- **Libra (lb):** Peso de un patrón de platino (libra estándar) con una masa de 0.45359243 kilogramos (kg), ubicada al nivel del mar y a 45° de latitud para una fuerza de 1 lb debido a la gravedad terrestre.

Estas unidades forman un sistema gravitacional, no absoluto, debido a su dependencia de la gravedad terrestre.


***Uso de la Libra como Unidad de Masa:***
- La libra estándar se usa como unidad de masa en comercio, pero no en cálculos de ingeniería.
- En ingeniería, la consistencia de unidades requiere una masa que, bajo una fuerza de 1 lb, tenga una aceleración de 1 pie/s², llamada **slug**.
  $$
  F = ma \quad 1 \, \text{lb} = (1 \, \text{slug}) \left(1 \, \text{pie} / \text{s}^2\right)
  $$
  $$
  1 \, \text{slug} = \frac{1 \, \text{lb}}{1 \, \text{pie/s}^2} = 1 \, \text{lb} \cdot \text{s}^2 / \text{pie}
  $$
- El **slug** es 32.2 veces la masa de una libra estándar.

***Ventajas y Uso en Ingeniería:***
- En estática, el peso se mide en libras.
- En dinámica, donde intervienen fuerzas, masas y aceleraciones, la masa (m) se expresa en slugs cuando el peso (W) está en libras:
  $$
  m = \frac{W}{g} \quad \text{donde} \quad g = 32.2 \, \text{pies}/\text{s}^2
  $$

***Otras Unidades Comunes:***
- **Milla (mi):** 5280 pies
- **Pulgada (in):** $\frac{1}{12}$ pie
- **Kilolibra (kip):** 1000 libras
- **Tonelada:** 2000 libras, convertida a slugs en cálculos de ingeniería.

***Conversión de Unidades:***
La conversión de unidades en el sistema estadounidense puede ser más complicada que en el SI. Ejemplo de conversión de velocidad:
- **Velocidad:** 30 mi/h a pies/s:
  $$
  v = 30 \, \frac{\text{mi}}{\text{h}} \times \frac{5280 \, \text{pies}}{1 \, \text{mi}} \times \frac{1 \, \text{h}}{3600 \, \text{s}}
  $$
  $$
  v = 44 \, \frac{\text{pies}}{\text{s}}
  $$


#### Conversión de Unidades entre el Sistema Internacional (SI) y el Sistema de Unidades de Uso Común en Estados Unidos

- ***Unidades de Longitud:*** Para convertir unidades de longitud entre el sistema estadounidense y el SI:
  - **Pie (ft):**
  $$  1 \text{ pie} = 0.3048 \, \text{m}  $$
  - **Milla (mi):**
  $$  1 \text{ mi} = 5280 \text{ pies} = 5280 \times 0.3048 \, \text{m} = 1609 \, \text{m} = 1.609 \, \text{km}  $$
  - **Pulgada (in):**
  $$  1 \text{ pulg} = \frac{1}{12} \text{ pie} = \frac{1}{12} \times 0.3048 \, \text{m} = 0.0254 \, \text{m} = 25.4 \, \text{mm}  $$

- ***Unidades de Fuerza:*** Para convertir unidades de fuerza:
  - **Libra (lb):**
  $$  1 \text{ lb} = (0.4536 \, \text{kg}) \times (9.807 \, \text{m/s}^2) = 4.448 \, \text{N}  $$

- ***Unidades de Masa:*** La masa en el sistema estadounidense se mide en slugs, que es una unidad derivada. Para convertir:
  - **Slug:**
  $$1 \text{ slug} = 1 \text{ lb} \cdot \text{s}^2 / \text{pie} = \frac{1 \text{ lb}}{1 \text{ pie/s}^2} = \frac{4.448 \text{ N}}{0.3048 \text{ m/s}^2} = 14.59 \, \text{N} \cdot \text{s}^2 / \text{m} = 14.59 \, \text{kg}$$
  - **Libra masa:**
  $$1 \text{ libra masa} = 0.4536 \, \text{kg}$$

- ***Conversión de Unidades Derivadas:*** Para convertir unidades derivadas:
  - **Momento de una Fuerza:**
  $$M = 47 \text{ lb} \cdot \text{ pulg} = 47 \times 4.448 \text{ N} \times 25.4 \text{ mm} = 5310 \text{ N} \cdot \text{ mm} = 5.31 \text{ N} \cdot \text{ m}$$

Para convertir de unidades del SI a unidades de uso común en Estados Unidos:
  - **Momento de una Fuerza:**
  $$M = 40 \text{ N} \cdot \text{ m} = 40 \text{ N} \cdot \text{ m} \times \left(\frac{1 \text{ lb}}{4.448 \text{ N}}\right) \times \left(\frac{1 \text{ pie}}{0.3048 \text{ m}}\right) = 29.5 \text{ lb} \cdot \text{ pie}$$
