### Introducción

En la Geometría Analítica Plana, el foco está en puntos situados en un plano, lo que limita el estudio a dos dimensiones y restringe la exploración de formas en el espacio tridimensional. Para superar esta limitación y estudiar figuras tridimensionales, se introduce una dimensión adicional, $ z $, que representa la distancia perpendicular de un punto $ P $ a un plano coordenado. Este enfoque transforma el sistema bidimensional en un sistema de coordenadas tridimensional, donde el sistema bidimensional es un caso especial del tridimensional si la dimensión $ z $ es cero. Así, una relación espacial tridimensional puede reducirse a su contraparte bidimensional asignando el valor cero a $ z $.

En contraste con la Geometría Analítica Plana, que utiliza ecuaciones con dos variables, la Geometría Analítica del Espacio generalmente involucra tres variables, añadiendo una complejidad analítica mayor y demandando una habilidad más avanzada para visualizar figuras en el espacio tridimensional.

### Sistema de coordenadas rectangulares en el espacio

El sistema de coordenadas rectangulares en el espacio, fundamental en la Geometría Analítica del Espacio, utiliza tres planos perpendiculares que se intersectan en un punto común, denominado origen ($O$). Este sistema permite designar las posiciones de los puntos en el espacio tridimensional. Los planos se conocen como planos coordenados y sus líneas de intersección como ejes coordenados. Los ejes, marcados como $ X X', Y Y', Z Z' $, pueden seguir un sistema de mano derecha o de mano izquierda, siendo el primero más común en este contexto.

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/CoordSystem05.PNG?raw=true" width="200" />
</p>

Cada eje es una recta dirigida con una dirección positiva indicada. Los planos coordenados, nombrados según los ejes que contienen (por ejemplo, el plano $XY$ contiene los ejes $X$ y $Y$), dividen el espacio en ocho regiones llamadas octantes. La posición de cualquier punto $ P $ en el espacio se determina a través de la intersección de planos paralelos a los planos coordenados con los ejes, formando un paralelepípedo recto. Las coordenadas del punto $ P $, expresadas como $ (x, y, z) $, se miden a partir del origen y pueden ser positivas o negativas dependiendo de su ubicación con respecto a la dirección positiva de los ejes.

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Punto3D02.PNG?raw=true" width="200" />
</p>

Cada punto en el espacio tiene una única terna de coordenadas en un sistema coordenado rectangular fijo y, a la inversa, cada terna de coordenadas representa un único punto en el espacio. Las coordenadas de un punto deben seguir un orden específico para indicar a lo largo de qué eje se mide cada una. Por tanto, existe una correspondencia biunívoca entre cada punto del espacio y una terna ordenada de números reales.

### Distancia entre dos puntos en el espacio

La distancia entre dos puntos en el espacio tridimensional se refiere a la longitud del segmento de línea más corto que conecta estos puntos. Este concepto se basa en el [Teorema de Pitágoras](https://study.com/academy/lesson/using-the-pythagorean-theorem-to-solve-3d-problems.html), que se extiende a tres dimensiones.

Consideremos dos puntos en el espacio tridimensional, $ P_1(x_1, y_1, z_1) $ y $ P_2(x_2, y_2, z_2) $. Estos puntos están definidos por sus coordenadas cartesianas, donde $ x $, $ y $ y $ z $ representan sus posiciones a lo largo de los ejes cartesianos.

La distancia $ d $ entre los puntos $ P_1 $ y $ P_2 $ se calcula mediante la siguiente fórmula, que es una extensión del Teorema de Pitágoras en tres dimensiones:

$$ d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2} $$

Donde:
- $ (x_2 - x_1) $, $ (y_2 - y_1) $ y $ (z_2 - z_1) $ son las diferencias de las coordenadas $ x $, $ y $, y $ z $ de los puntos $ P_2 $ y $ P_1 $, respectivamente.

Geométricamente, esta fórmula se interpreta como la hipotenusa de un triángulo rectángulo en tres dimensiones, donde las diferencias de las coordenadas forman los lados perpendiculares del triángulo.

#### División de un segmento en el espacio en una razón dada

Ahora consideraremos la división de un segmento dado en el espacio en una razón dada. Esto es, simplemente, una ampliación del problema análogo en el plano, que se ha estudiado en el capítulo anterior de la ecuación de la Recta.

***Teorema:*** Si $\mathrm{P}_1\left(\mathrm{x}_1, \mathrm{y}_1, \mathrm{z}_1\right)$ y $\mathrm{P}_2\left(\mathrm{x}_2, \mathrm{y}_2, \mathrm{z}_2\right)$ son los extremos de un segmento dirigido $\mathrm{P}_1 \mathrm{P}_2$, las coordenadas $(\mathrm{x}, \mathrm{y}, \mathrm{z}$ ) de un punto $\mathrm{P}$ que divide a este segmento en la razón $\mathrm{r}=\overline{\mathrm{P}_1 \mathrm{P}}: \overline{\mathrm{PP}_2}$ son
$$
\mathrm{x}=\frac{\mathrm{x}_1+\mathrm{rx}_2}{1+\mathrm{r}}, \quad \mathrm{y}=\frac{\mathrm{y}_1+\mathrm{ry_{2 }}}{1+\mathrm{r}}, \quad \mathrm{z}=\frac{\mathrm{z}_1+\mathrm{rz}_2}{1+\mathrm{r}}, \quad(\mathrm{r} \neq-1)
$$

Igual que en el capítulo anterior, se pueden encontrar formulaciones para la razón, $r$ em función de las coordenadas de los puntos $\mathrm{P}_1$ y $\mathrm{P}_2$. Basta despejar la razón $\mathrm{r}$ de las ecuaciones anteriores.

### Cosenos directores

En geometría analítica, los cosenos directores son una forma de representar la orientación de una recta en el espacio tridimensional. Proporcionan una medida angular de la dirección de la recta en relación con los ejes coordenados. En geometría analítica plana, la dirección de una recta en el plano se determina por medio de su ángulo de inclinación o de su pendiente. En este capítulo veremos cómo se determina la dirección de una recta en el espacio.

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Cosenos_directores01.PNG?raw=true" width="350" />
</p>

De la figura en $(a)$, sea $l$ una recta que pasa por los puntos $P_1\left(x_1, y_1, z_1\right)$ y $P_2\left(x_2, y_2, z_2\right)$. Primero consideraremos el caso en que $l$ tiene el sentido indicado en la figura. Por cada uno de los puntos $P_1$ y $P_2$, hagamos pasar planos paralelos a los coordenados, formando asi un paralelepípedo recto rectangular cuya diagonal es $P_1 P_2$, y cuyas aristas paralelas a los ejes $X, Y$ y $Z$ son, respectivamente, $P_1 V_1, P_1 V_2$ y $P_1 V_3$. Si cada arista tiene el mismo sentido que el eje a que es paralela, los ángulos directores son
$$\alpha=\text { ángulo } P_2 P_1 V_1$$  
$$\beta=\text { ángulo } P_2 P_1 V_2$$  
$$\gamma=\text { ángulo } P_2 P_1 V_3$$

Ahora consideremos figs. $(b)$,$(c)$ y $(d)$, los tres triángulos rectángulos formados por los dos puntos $P_1, P_2$ y cada uno de los vértices $V_1, V_2$ y $V_3$. Para cada uno de estos triángulos sea $d=\left|\overline{P_1 P_2}\right|$, en que $d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2}$, es la distancia en 3D entre los dos puntos. También,
$$
\overline{P_1 V_1}=x_2-x_1, \quad \overline{P_1 V_2}=y_2-y_1, \quad \overline{P_1 V_3}=z_2-z_1 .
$$

Por tanto, de los tres triángulos, tenemos, para los cosenos directores,
$$
\cos \alpha=\frac{x_2-x_1}{d}, \quad \cos \beta=\frac{y_2-y_1}{d}, \quad \cos \gamma=\frac{z_2-z_1}{d} .
$$

Si la recta $l$ se considera dirigida en el sentido de $P_2$ a $P_1$, entonces los tres cosenos directores son
$$
\cos \alpha=\frac{x_1-x_2}{d}, \quad \cos \beta=\frac{y_1-y_2}{d}, \quad \cos \gamma=\frac{z_1-z_2}{d} .
$$

***Teorema:*** Los cosenos directores de la recta determinada por los dos puntos $\mathrm{P}_1\left(\mathrm{x}_1, \mathrm{y}_1, \mathrm{z}_1\right)$ y $\mathrm{P}_2\left(\mathrm{x}_2, \mathrm{y}_2, \mathrm{z}_2\right)$, y dirigida en el sentido de $\mathrm{P}_1$ a $\mathrm{P}_2$, son
$$
\cos \alpha=\frac{\mathrm{x}_2-\mathrm{x}_1}{\mathrm{~d}}, \quad \cos \beta=\frac{\mathrm{y}_2-\mathrm{y}_1}{\mathrm{~d}}, \quad \cos \gamma=\frac{\mathrm{z}_2-\mathrm{z}_1}{\mathrm{~d}},
$$
siendo $\mathrm{d}$ la distancia entre $\mathrm{P}_1$ y $\mathrm{P}_2$.

Elevando al cuadrado cada uno de los cosenos directores, se llega a:

$$\cos^2(\alpha) + \cos^2(\beta) + \cos^2(\gamma) = 1$$

#### Ejemplos

***1.*** Determinar los cosenos directores del vector $\boldsymbol{A}=(5,7,-3)$.

Como $\boldsymbol{A}=\left(\mathrm{A}_{\mathrm{x}}, \mathrm{A}_{\mathrm{y}}, \mathrm{A}_{\mathrm{z}}\right)=(5,7,-3)$ entonces:
$$
A_x=5 \quad ; \quad A_y=7 \quad ; \quad A_z=-3
$$

Ahora, los cosenos directores vendrán dados por:
$$
\cos \alpha=\frac{A_x}{|\vec{A}|} \quad ; \quad \cos \beta=\frac{A_y}{|\vec{A}|} \quad ; \quad \cos \gamma=\frac{A_z}{|\vec{A}|}
$$

Calculemos el módulo (o magnitud) de $\boldsymbol{A}$ :
$$
\begin{aligned}
|\vec{A}| & =\sqrt{\left(A_x\right)^2+\left(A_y\right)^2+\left(A_z\right)^2}= \\
& =\sqrt{(5)^2+(7)^2+(-3)^2}= \\
& =\sqrt{25+49+9}= \\
& =\sqrt{83}
\end{aligned}
$$

Por lo tanto:
$$
\cos \alpha=\frac{5}{|\sqrt{83}|} \quad ; \quad \cos \beta=\frac{7}{|\sqrt{83}|} \quad ; \quad \cos \gamma=\frac{-3}{|\sqrt{83}|}
$$

***2.*** Determine los cosenos directores del vector $\boldsymbol{A}=(1,-2,4)$ y compruebe que la suma de los cuadrados de todos los cosenos directores es igual a uno.

Como $\boldsymbol{A}=\left(A_x, A_{y^{\prime}} A_z\right)=(1,-2,4)$ entonces:
$$
A_x=1 \quad ; \quad A_y=-2 \quad ; \quad A_z=4
$$

Los cosenos directores vendrán dados por:
$$
\cos \alpha=\frac{A_x}{|\vec{A}|} \quad ; \quad \cos \beta=\frac{A_y}{|\vec{A}|} \quad ; \quad \cos \gamma=\frac{A_z}{|\vec{A}|}
$$

Calculemos el módulo (o magnitud) de $\boldsymbol{A}$ :
$$
\begin{aligned}
|\vec{A}| & =\sqrt{\left(A_x\right)^2+\left(A_y\right)^2+\left(A_z\right)^2}= \\
& =\sqrt{(1)^2+(-2)^2+(4)^2}= \\
& =\sqrt{1+4+16}= \\
& =\sqrt{21}
\end{aligned}
$$

Por lo tanto:
$$
\begin{aligned}
& \cos \alpha=\frac{1}{|\sqrt{21}|} \quad ; \quad \cos \beta=\frac{-2}{|\sqrt{21}|} \quad ; \quad \cos \gamma=\frac{4}{|\sqrt{21}|} \\
& \cos \alpha=1 / \sqrt{21} \\
& \cos \beta=-2 / \sqrt{21} \\
& \cos \gamma=4 / \sqrt{ } 21
\end{aligned}
$$

Ahora, debemos comprobar que:
$$
\cos ^2 \alpha+\cos ^2 \beta+\cos ^2 \gamma=1
$$

Por lo tanto:
$$
\cos \left(\frac{1}{\sqrt{21}}\right)^2+\cos \left(\frac{-2}{\sqrt{21}}\right)^2+\cos \left(\frac{4}{\sqrt{21}}\right)^2=\frac{1}{21}+\frac{4}{21}+\frac{16}{21}=\frac{21}{21}=1
$$

Que es lo que deseábamos comprobar.

$$

$$

Los cosenos directores vendrán dados por:
$$

$$

Calculemos el módulo (o magnitud) de $\boldsymbol{A}$ :
$$
\begin{aligned}
|\vec{A}| & =\sqrt{\left(A_x\right)^2+\left(A_y\right)^2}= \\
& =\sqrt{(-4)^2+(3)^2}= \\

& =\sqrt{25}= \\

\end{aligned}
