### Introducción

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/CoordSystem04.PNG?raw=true" width="350" />
</p>

En la Geometría Analítica Plana, el foco está en puntos situados en un plano, lo que limita el estudio a dos dimensiones y restringe la exploración de formas en el espacio tridimensional. Para superar esta limitación y estudiar figuras tridimensionales, se introduce una dimensión adicional, $ z $, que representa la distancia perpendicular de un punto $ P $ a un plano coordenado. Este enfoque transforma el sistema bidimensional en un sistema de coordenadas tridimensional, donde el sistema bidimensional es un caso especial del tridimensional si la dimensión $ z $ es cero. Así, una relación espacial tridimensional puede reducirse a su contraparte bidimensional asignando el valor cero a $ z $.

En contraste con la Geometría Analítica Plana, que utiliza ecuaciones con dos variables, la Geometría Analítica del Espacio generalmente involucra tres variables, añadiendo una complejidad analítica mayor y demandando una habilidad más avanzada para visualizar figuras en el espacio tridimensional.

### Sistema de coordenadas rectangulares en el espacio

El sistema de coordenadas rectangulares en el espacio, fundamental en la Geometría Analítica del Espacio, utiliza tres planos perpendiculares que se intersectan en un punto común, denominado origen ($O$). Este sistema permite designar las posiciones de los puntos en el espacio tridimensional. Los planos se conocen como planos coordenados y sus líneas de intersección como ejes coordenados. Los ejes, marcados como $ X X', Y Y', Z Z' $, pueden seguir un sistema de mano derecha o de mano izquierda, siendo el primero más común en este contexto.

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/CoordSystem05.PNG?raw=true" width="200" />
</p>

Cada eje es una recta dirigida con una dirección positiva indicada. Los planos coordenados, nombrados según los ejes que contienen (por ejemplo, el plano $XY$ contiene los ejes $X$ y $Y$), dividen el espacio en ocho regiones llamadas octantes. La posición de cualquier punto $ P $ en el espacio se determina a través de la intersección de planos paralelos a los planos coordenados con los ejes, formando un paralelepípedo recto. Las coordenadas del punto $ P $, expresadas como $ (x, y, z) $, se miden a partir del origen y pueden ser positivas o negativas dependiendo de su ubicación con respecto a la dirección positiva de los ejes.

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Punto3D02.PNG?raw=true" width="200" />
</p>

Cada punto en el espacio tiene una única terna de coordenadas en un sistema coordenado rectangular fijo y, a la inversa, cada terna de coordenadas representa un único punto en el espacio. Las coordenadas de un punto deben seguir un orden específico para indicar a lo largo de qué eje se mide cada una. Por tanto, existe una correspondencia biunívoca entre cada punto del espacio y una terna ordenada de números reales.

## Distancia entre dos puntos en el espacio

La distancia entre dos puntos en el espacio tridimensional se refiere a la longitud del segmento de línea más corto que conecta estos puntos. Este concepto se basa en el [Teorema de Pitágoras](https://study.com/academy/lesson/using-the-pythagorean-theorem-to-solve-3d-problems.html), que se extiende a tres dimensiones.

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/pythagoras-3d-d.svg?raw=true" width="200" />
</p>

<p style="text-align: center;">

</p>


Consideremos dos puntos en el espacio tridimensional, $ P_1(x_1, y_1, z_1) $ y $ P_2(x_2, y_2, z_2) $. Estos puntos están definidos por sus coordenadas cartesianas, donde $ x $, $ y $ y $ z $ representan sus posiciones a lo largo de los ejes cartesianos.

La distancia $ d $ entre los puntos $ P_1 $ y $ P_2 $ se calcula mediante la siguiente fórmula, que es una extensión del Teorema de Pitágoras en tres dimensiones:

$$ d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2} $$

Donde:
- $ (x_2 - x_1) $, $ (y_2 - y_1) $ y $ (z_2 - z_1) $ son las diferencias de las coordenadas $ x $, $ y $, y $ z $ de los puntos $ P_2 $ y $ P_1 $, respectivamente.

Geométricamente, esta fórmula se interpreta como la hipotenusa de un triángulo rectángulo en tres dimensiones, donde las diferencias de las coordenadas forman los lados perpendiculares del triángulo.


### Ejemplo

Calcular la distancia desde el origen hasta el punto $ Q(3, -2, 5) $ en un sistema de coordenadas tridimensional.

La fórmula para calcular la distancia $ d $ entre dos puntos $ P_1(x_1, y_1, z_1) $ y $ P_2(x_2, y_2, z_2) $ en el espacio 3D es:

$$ d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2} $$

Dado que uno de los puntos es el origen $ O(0, 0, 0) $, la fórmula se simplifica a:

$$ d = \sqrt{x^2 + y^2 + z^2} $$

Para nuestro punto $ Q(3, -2, 5) $, la distancia al origen será:

$$ d = \sqrt{3^2 + (-2)^2 + 5^2} $$

La distancia del origen al punto $Q(3, -2, 5)$ es $d \approx 6.16$ unidades.

***Teorema:*** Si $\mathrm{P}_1\left(\mathrm{x}_1, \mathrm{y}_1, \mathrm{z}_1\right)$ y $\mathrm{P}_2\left(\mathrm{x}_2, \mathrm{y}_2, \mathrm{z}_2\right)$ son los extremos de un segmento dirigido $\mathrm{P}_1 \mathrm{P}_2$, las coordenadas $(\mathrm{x}, \mathrm{y}, \mathrm{z}$ ) de un punto $\mathrm{P}$ que divide a este segmento en la razón $\mathrm{r}=\overline{\mathrm{P}_1 \mathrm{P}}: \overline{\mathrm{PP}_2}$ son

$$

## Cosenos  y Números directores

### Cosenos directores

En geometría analítica, los cosenos directores son una forma de representar la orientación de una recta en el espacio tridimensional. Proporcionan una medida angular de la dirección de la recta en relación con los ejes coordenados. En geometría analítica plana, la dirección de una recta en el plano se determina por medio de su ángulo de inclinación o de su pendiente. En este capítulo veremos cómo se determina la dirección de una recta en el espacio.

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Cosenos_directores01.PNG?raw=true" width="350" />
</p>

De la figura en $(a)$, sea $l$ una recta que pasa por los puntos $P_1\left(x_1, y_1, z_1\right)$ y $P_2\left(x_2, y_2, z_2\right)$. Primero consideraremos el caso en que $l$ tiene el sentido indicado en la figura. Por cada uno de los puntos $P_1$ y $P_2$, hagamos pasar planos paralelos a los coordenados, formando asi un paralelepípedo recto rectangular cuya diagonal es $P_1 P_2$, y cuyas aristas paralelas a los ejes $X, Y$ y $Z$ son, respectivamente, $P_1 V_1, P_1 V_2$ y $P_1 V_3$. Si cada arista tiene el mismo sentido que el eje a que es paralela, los ángulos directores son
$$\alpha=\text { ángulo } P_2 P_1 V_1$$  
$$\beta=\text { ángulo } P_2 P_1 V_2$$  
$$\gamma=\text { ángulo } P_2 P_1 V_3$$

Ahora consideremos figs. $(b)$,$(c)$ y $(d)$, los tres triángulos rectángulos formados por los dos puntos $P_1, P_2$ y cada uno de los vértices $V_1, V_2$ y $V_3$. Para cada uno de estos triángulos sea $d=\left|\overline{P_1 P_2}\right|$, en que $d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2}$, es la distancia en 3D entre los dos puntos. También,
$$
\overline{P_1 V_1}=x_2-x_1, \quad \overline{P_1 V_2}=y_2-y_1, \quad \overline{P_1 V_3}=z_2-z_1 .
$$

Por tanto, de los tres triángulos, tenemos, para los cosenos directores,
$$
\cos \alpha=\frac{x_2-x_1}{d}, \quad \cos \beta=\frac{y_2-y_1}{d}, \quad \cos \gamma=\frac{z_2-z_1}{d} .
$$

Si la recta $l$ se considera dirigida en el sentido de $P_2$ a $P_1$, entonces los tres cosenos directores son
$$
\cos \alpha=\frac{x_1-x_2}{d}, \quad \cos \beta=\frac{y_1-y_2}{d}, \quad \cos \gamma=\frac{z_1-z_2}{d} .
$$

***Teorema:*** Los cosenos directores de la recta determinada por los dos puntos $\mathrm{P}_1\left(\mathrm{x}_1, \mathrm{y}_1, \mathrm{z}_1\right)$ y $\mathrm{P}_2\left(\mathrm{x}_2, \mathrm{y}_2, \mathrm{z}_2\right)$, y dirigida en el sentido de $\mathrm{P}_1$ a $\mathrm{P}_2$, son
$$
\cos \alpha=\frac{\mathrm{x}_2-\mathrm{x}_1}{\mathrm{~d}}, \quad \cos \beta=\frac{\mathrm{y}_2-\mathrm{y}_1}{\mathrm{~d}}, \quad \cos \gamma=\frac{\mathrm{z}_2-\mathrm{z}_1}{\mathrm{~d}},
$$
siendo $\mathrm{d}$ la distancia entre $\mathrm{P}_1$ y $\mathrm{P}_2$.

Elevando al cuadrado cada uno de los cosenos directores, se llega a:

$$\cos^2(\alpha) + \cos^2(\beta) + \cos^2(\gamma) = 1$$

### Ejemplos

***1.*** Determinar los cosenos directores del vector $\boldsymbol{A}=(5,7,-3)$.

Como $\boldsymbol{A}=\left(\mathrm{A}_{\mathrm{x}}, \mathrm{A}_{\mathrm{y}}, \mathrm{A}_{\mathrm{z}}\right)=(5,7,-3)$ entonces:
$$
A_x=5 \quad ; \quad A_y=7 \quad ; \quad A_z=-3
$$

Ahora, los cosenos directores vendrán dados por:
$$
\cos \alpha=\frac{A_x}{|\vec{A}|} \quad ; \quad \cos \beta=\frac{A_y}{|\vec{A}|} \quad ; \quad \cos \gamma=\frac{A_z}{|\vec{A}|}
$$

Calculemos el módulo (o magnitud) de $\boldsymbol{A}$ :
$$
\begin{aligned}
|\vec{A}| & =\sqrt{\left(A_x\right)^2+\left(A_y\right)^2+\left(A_z\right)^2}= \\
& =\sqrt{(5)^2+(7)^2+(-3)^2}= \\
& =\sqrt{25+49+9}= \\
& =\sqrt{83}
\end{aligned}
$$

Por lo tanto:
$$
\cos \alpha=\frac{5}{|\sqrt{83}|} \quad ; \quad \cos \beta=\frac{7}{|\sqrt{83}|} \quad ; \quad \cos \gamma=\frac{-3}{|\sqrt{83}|}
$$

***2.*** Determine los cosenos directores del vector $\boldsymbol{A}=(1,-2,4)$ y compruebe que la suma de los cuadrados de todos los cosenos directores es igual a uno.

$$
A_x=1 \quad ; \quad A_y=-2 \quad ; \quad A_z=4
$$

Los cosenos directores vendrán dados por:
$$
\cos \alpha=\frac{A_x}{|\vec{A}|} \quad ; \quad \cos \beta=\frac{A_y}{|\vec{A}|} \quad ; \quad \cos \gamma=\frac{A_z}{|\vec{A}|}

Calculemos el módulo (o magnitud) de $\boldsymbol{A}$ :
$$
\begin{aligned}
|\vec{A}| & =\sqrt{\left(A_x\right)^2+\left(A_y\right)^2+\left(A_z\right)^2}= \\
& =\sqrt{(1)^2+(-2)^2+(4)^2}= \\
& =\sqrt{1+4+16}= \\
& =\sqrt{21}
\end{aligned}
$$

Por lo tanto:
$$
\begin{aligned}
& \cos \alpha=\frac{1}{\sqrt{21}} \quad ; \quad \cos \beta=\frac{-2}{\sqrt{21}} \quad ; \quad \cos \gamma=\frac{4}{\sqrt{21}} 
\end{aligned}
$$

Ahora, debemos comprobar que:
$$
\cos ^2 \alpha+\cos ^2 \beta+\cos ^2 \gamma=1

Por lo tanto:

$$

Que es lo que deseábamos comprobar.

### Números directores de una recta en el espacio

En lugar de los cosenos directores de una recta $l$ conviene, a veces, emplear tres números reales, llamados *números directores* de $l$, que sean proporcionales a sus cosenos directores. Así, $a, b$ y $c$ son los números directores de una recta $l$, siempre que

$$\frac{a}{\cos \alpha}=\frac{b}{\cos \beta}=\frac{c}{\cos \gamma}, \quad\quad (1)$$  

en donde $\cos \alpha, \cos \beta$ y $\cos \gamma$ son los cosenos directores de $l$.

Es conveniente introducir una notación especial para ellos. Si tres números reales cualesquiera, $a, b$ y $c$, representan los números directores de una recta, indicaremos esto encerrándolos entre paréntesis rectangulares, asi : $[a, b, c]$. Los paréntesis rectangulares sirven para distinguir los números directores de una recta de las coordenadas de un punto que se encierran en paréntesis ordinarios.

Los cosenos directores de una recta pueden determinarse fácilmente a partir de sus números directores. En efecto, igualemos cada una de las razones de la ecuación anterior a algún número $k$ diferente de cero, de modo que

$$a=k \cos \alpha, \quad b=k \cos \beta, \quad c=k \cos \gamma . \quad\quad (2)$$

Si elevamos al cuadrado ambos miembros de las ecuaciones en $(2)$, y sumamos, obtenemos

$$a^2+b^2+c^2=k^2\left(\cos ^2 \alpha+\cos ^2 \beta+\cos ^2 \gamma\right),$$

se reduce a

$$a^2+b^2+c^2=k^2 $$

de manera que

$$k= \pm \sqrt{a^2+b^2+c^2}$$

Por tanto, de las ecuaciones $(2)$, tenemos el siguiente teorema:

***Teorema:*** Si $[\mathrm{a}, \mathrm{b}, \mathrm{c}]$ son los números directores de una recta, sus cosenos directores son

$$
\begin{gathered}
\cos \alpha= \pm \frac{a}{\sqrt{a^2+b^2+c^2}}, \\
\cos \beta= \pm \frac{b}{\sqrt{a^2+b^2+c^2}}, \\
\cos \gamma= \pm \frac{c}{\sqrt{a^2+b^2+c^2}},
\end{gathered}
$$

en donde se escoge el signo superior o el inferior según que la recta esté dirigida en un sentido o en el sentido opuesto.

### Ejemplos

***1.*** Los números directores de una recta $l$ son $[2,-2,-1]$. Hallar los cosenos directores de $l$ si la recta está dirigida de tal manera que el ángulo $\beta$ es agudo.

- Solución. Por el teorema 5 anterior, los cosenos directores de $l$, cuando la recta no está dirigida, son

$$\cos \alpha= \pm \frac{2}{\sqrt{2^2+(-2)^2+(-1)^2}}= \pm 2 / 3, \cos \beta=\mp 2 / 3, \cos \gamma=\mp 1 / 3$$

Como $l$ está dirigida de tal manera que $\beta$ es agudo, $\cos \beta$ es positivo. Por tanto, tomando los signos inferiores para los cosenos directores, tendremos

$$\cos \alpha=-2 / 3, \quad \cos \beta=2 / 3, \quad \cos \gamma=1/3 .$$

## Angulo formado por dos rectas dirigidas en el espacio

### Introducción

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/AnguloRectasEspacio.PNG?raw=true" width="200" />
</p>

Sean $l_1^{\prime}$ y $l_2^{\prime}$ dos rectas trazadas por el origen y paralelas, y del mismo sentido, a $l_1$ y $l_2$, respectivamente. Por definición, el ángulo formado por las rectas dirigidas $l_1$ y $l_2$ es el ángulo $\theta$. Sea $P_1\left(x_1, y_1, z_1\right)$ un punto cualquiera, distinto del origen, sobre $l^{\prime}{ }_1$, y $P_2\left(x_2, y_2, z_2\right)$ otro punto cualquiera, distinto del origen sobre $l_2^{\prime}$. También, sea

$\left|\overline{O P_1}\right|=d_1,\left|\overline{O P_2}\right|=d_2$ y $\left|\overline{P_1 P_2}\right|=d$. Por la ley de los cosenos, tenemos, para el triángulo $O P_1 P_2$,

$$\cos \theta=\frac{d_1{ }^2+d_2{ }^2-d^2}{2 d_1 d_2}  \quad \quad (1)$$

Por la distancia de un segmento de recta en el espacio desde el origen, tenemos

$$
\begin{gathered}
d_1{ }^2=x_1{ }^2+y_1{ }^2+z_1{ }^2, \\
d_2{ }^2=x_2{ }^2+y_2{ }^2+z_2{ }^2, \\
d^2=\left(x_2-x_1\right)^2+\left(y_2-y_1\right)^2+\left(z_2-z_1\right)^2 .
\end{gathered}
$$

Sustituyendo estos valores en el numerador del segundo miembro de la ecuación $(1)$, y simplificamos, obtenemos

$$\cos \theta=\frac{x_2 x_2+y_1 y_2+z_1 z_2}{d_1 d_2}  \quad \quad (2)$$

Sean $\alpha_1, \beta_1, \gamma_1$ los ángulos directores de $l_1$ y, por tanto, de $l_1^{\prime}$, y $\alpha_2, \beta_2, \gamma_2$ los ángulos directores de $l_2$ y, por tanto, de $l_2^{\prime}$, tenemos:

$$
\begin{aligned}
\cos \alpha_1=\frac{x_1}{d_1}, \quad \cos \beta_1=\frac{y_1}{d_1}, \quad \cos \gamma_1=\frac{z_1}{d_1} \\
\mathbf{y} \quad \cos \alpha_2=\frac{x_2}{d_2}, \quad \cos \beta_2=\frac{y_2}{d_2}, \quad \cos \gamma_2=\frac{z_1}{d_2} .
\end{aligned}
$$

Sustituyendo estos valores en la ecuación $(2)$, obtenemos la relación buscada

$$\cos \theta=\cos \alpha_1 \cos \alpha_2+\cos \beta_1 \cos \beta_2+\cos \gamma_1 \cos \gamma_2 \quad \quad (3)$$

Esta igualdad nos dice:

- ***Teorema:*** El ángulo $\theta$ formado por dos rectas dirigidas cualesquiera en el espacio, cuyos ángulos directores son $\alpha_1, \beta_1, \gamma_1 y$ $\alpha_2, \beta_2, \gamma_2$, respectivamente, se determina por la relación

$$\cos \theta=\cos \alpha_1 \cos \alpha_2+\cos \beta_1 \cos \beta_2+\cos \gamma_1 \cos \gamma_2$$

- Para que dos rectas sean paralelas y del mismo sentido es condición necesaria y suficiente que sus ángulos directores correspondientes sean iguales; para que sean paralelas $y$ de sentidos opuestos es necesario y suficiente que sus angulos directores correspondientes sean suplementarios.

- Para que dos rectas dirigidas sean perpendiculares es necesario $y$ suficiente que la suma de los productos de sus cosenos directores correspondientes sea igual a cero.

Ahora vamos a obtener estos resultados en función de los números directores de las dos rectas.

Sean $\left[a_1, b_1, c_1\right]$ y $\left[a_2, b_2, c_2\right]$ los números directores de las dos rectas $l_1$ y $l_2$, respectivamente. Por un teorema anterior, tenemos

$$
\begin{gathered}
\cos \alpha_1= \pm \frac{a_1}{\sqrt{a_1^2+b_1^2+c_1^2}}, \\
\cos \beta_1= \pm \frac{b_1}{\sqrt{a_1^2+b_1^2+c_1^2}} \\
\cos \gamma_1= \pm \frac{c_1}{\sqrt{a_1^2+b_1^2+c_1^2}},
\end{gathered}
$$

y

$$
\begin{gathered}
\cos \alpha_2= \pm \frac{a_2}{\sqrt{a_2^2+b_2^2+c_2^2}}, \\
\cos \beta_2= \pm \frac{b_2}{\sqrt{a_2^2+b_2^2+c_2^2}}, \\
\cos \gamma_2= \pm \frac{c_2}{\sqrt{a_2^2+b_2^2+c_2^2}}
\end{gathered}
$$


Sustituyendo estos valores en la relación del teorema anterior , obtenemos:

El angulo $\theta$ formado por dos rectas dirigidas cualesquiera en el espacio, cuyos números directores son $\left[\mathrm{a}_1, \mathrm{b}_1, \mathrm{c}_1\right]$ y $\left[\mathrm{a}_2, \mathrm{b}_2, \mathrm{c}_2\right]$, respectivamente, está determinado por la relación

$$
\cos \theta= \pm \frac{\mathrm{a}_1 \mathrm{a}_2+\mathrm{b}_1 \mathrm{b}_2+\mathrm{c}_1 \mathrm{c}_2}{\sqrt{\mathrm{a}_1^2+\mathrm{b}_1^2+\mathrm{c}_1^2} \sqrt{\mathrm{a}_2^2+\mathrm{b}_2^2+\mathrm{c}_2^2}} .
$$

***Nota:*** El doble signo indica que hay dos valores de $\theta$, suplementarios entre si. Un valor específico de $\theta$ puede obtenerse siempre considerando los dos sentidos de las rectas. Esto se ilustra en el ejemplo que veremos adelante.

Del teorema anterior se deducen los dos corolarios siguientes:

- ***Corolario 1:*** Para que dos rectas dirigidas sean paralelas es necesario y suficiente que sus números directores correspondientes sean proporcionales.

- ***Corolario 2:*** Para que dos rectas dirigidas sean perpendiculares es necesario y suficiente que la suma de los productos de sus números directores correspondientes sea igual a cero.

### Ejemplo

Hallar el área del triángulo cuyos vértices son los puntos $P_1(1,-1,2), P_2(4,5,-7)$ y $P_3(-1,2,1)$.

</p>

Sea el ángulo $P_2 P_1 P_3=\theta$, $\left|\overline{P_1 P_2}\right|=d_1$ y $\left|\overline{P_1 P_3}\right|=d_2$. El área del triángulo está dada por:

$$K=1 / 2 d_1 d_2 \sin \theta \quad \quad (4)$$

El sentido de los lados del ángulo $\theta$ correspondiente al vértice $P_1$ es el indicado en la figura. Para obtener los signos correctos de los cosenos directores de estos lados, restamos las coordenadas de $P_1$ de las coordenadas correspondientes de $P_2$ y $P_3$, por tanto, los números directores de $P_1 P_2$ son $[4-1,5+1,-7-2]$, o sea, $[3,6,-9]$ ó $[1,2,-3]$ y los de $P_1 P_3$ son $[-1-1,2+1,1-2]$, o sea, $[-2,3,-1]$.

Por tanto, tenemos

$$

### Números directores de una recta perpendicular a dos dadas. 

Sean $\left[a_1, b_1, c_1\right]$ y $\left[a_2, b_2, c_2\right]$ los números directores dados de dos rectas no paralelas, $l_1$ y $l_2$, respectivamente. Queremos determinar los números directores $[a, b, c]$ de una recta cualquiera $l$ perpendicular a ambas $l_1$ y $l_2$. Tal recta existe. En efecto, si $l_1$ y $l_2$ se cortan, $l$ puede representar una cualquiera de las rectas paralelas perpendiculares al plano determinado por $l_1$ y $l_2$. Si $l_1$ y $l_2$ se cruzan, entonces $l$ puede representar una cualquiera de las rectas perpendiculares al plano determinado por dos rectas que se cortan y son paralelas respectivamente a $l_1$ y $l_2$.

Como $l$ es perpendicular a $l_1$ y $l_2$, tenemos, por el corolario 2 del teorema anterior, las dos relaciones siguientes

$$
\left.\begin{array}{l}
a_1 a+b_1 b+c_1 c=0 \\
a_2 a+b_2 b+c_2 c=0
\end{array}\right\}
$$

El sistema (1) consta de dos ecuaciones con tres incógnitas, $a, b$ y $c$. Podemos resolver este sistema para dos cualesquiera de estas incógnitas en función de la tercera por la regla de Cramer siempre que el determinante del sistema sea diferente de cero. Este determinante puede ser uno cualquiera de los tres determinantes

$$
\left|\begin{array}{ll}
a_1 & b_1 \\
a_2 & b_2
\end{array}\right|,\left|\begin{array}{ll}
a_1 & c_1 \\
a_2 & c_2
\end{array}\right|,\left|\begin{array}{ll}
b_1 & c_1 \\
b_2 & c_2
\end{array}\right|
$$

Uno, por lo menos, de estos determinantes es diferente de cero. En efecto, si fueran todos nulos, tendríamos, respectivamente,

$$a_1 b_2=a_2 b_1, \quad a_1 c_2=a_2 c_1, \quad b_1 c_2=b_2 c_1,$$

de donde,

$$
\frac{a_1}{a_2}=\frac{b_1}{b_2}=\frac{c_1}{c_2},
$$

y por el corolario 1 del teorema 7, esta última relación implica que $l_1$ y $l_2$ sean paralelas, lo que contradice la hipótesis. Por tanto, podemos suponer que el primero de los determinantes (2) es diferente de cero, y resolver el sistema (1) para $a$ y $b$ en términos de $c$.

Esto nos da

$$
a=\frac{\left|\begin{array}{rr}
-c_1 c & b_1 \\
-c_2 c & b_2
\end{array}\right|}{\left|\begin{array}{rr}
a_1 & b_1 \\
a_2 & b_2
\end{array}\right|}=\frac{\left|\begin{array}{ll}
b_1 & c_1 \\
b_2 & c_2
\end{array}\right|}{\left|\begin{array}{ll}
a_1 & b_1 \\
a_2 & b_2
\end{array}\right|} c, \quad b=\frac{\left|\begin{array}{ll}
a_1 & -c_1 c \\
a_2 & -c_2 c
\end{array}\right|}{\left|\begin{array}{ll}
a_1 & b_1 \\
a_2 & b_2
\end{array}\right|}=\frac{\left|\begin{array}{ll}
c_1 & a_1 \\
c_2 & a_2
\end{array}\right|}{\left|\begin{array}{ll}
a_1 & b_1 \\
a_2 & b_2
\end{array}\right|} c .
$$

Ahora bien, $c$ no puede ser cero. Porque, si $c=0$, las últimas relaciones indican que $a$ y $b$ son ambas iguales a cero, lo que está en contradicción con el corolario 1 del teorema 5. Como los números directores de una recta no son únicos, podemos, por simplicidad, escoger el sistema en que $c=1$. Entonces los números directores de $l$ son

$$
a=\frac{\left|\begin{array}{ll}
b_1 & c_1 \\
b_2 & c_2
\end{array}\right|}{\left|\begin{array}{ll}
a_1 & b_1 \\
a_2 & b_2
\end{array}\right|}, \quad b=\frac{\left|\begin{array}{ll}
c_1 & a_1 \\
c_2 & a_2
\end{array}\right|}{\left|\begin{array}{ll}
a_1 & b_1 \\
a_2 & b_2
\end{array}\right|}, \quad c=1 .
$$

Para mayor simplicidad, multipliquemos este sistema por el denominador que es diferente de cero. Esto nos da, finalmente, el sistema de números directores

$$
a=\left|\begin{array}{ll}
b_1 & c_1 \\
b_2 & c_2
\end{array}\right|, \quad b=\left|\begin{array}{ll}
c_1 & a_1 \\
c_2 & a_2
\end{array}\right|, \quad c=\left|\begin{array}{ll}
a_1 & b_1 \\
a_2 & b_2
\end{array}\right| .
$$

Este resultado nos dice:

***Teorema:*** Si $\left[\mathrm{a}_1, \mathrm{b}_1, \mathrm{c}_1\right] y\left[\mathrm{a}_2, \mathrm{b}_2, \mathrm{c}_2\right]$ son los números directores dados de dos rectas no paralelas, $\mathrm{l}_1 y \mathrm{l}_2$, respectivamente, los números directores $[\mathrm{a}, \mathrm{b}, \mathrm{c}]$ de cualquier recta 1 perpendicular a ambas $1_1 y l_2$ están dados por los determinantes

$$
a=\left|\begin{array}{ll}
b_1 & c_1 \\
b_2 & c_2
\end{array}\right|, \quad b=\left|\begin{array}{ll}
c_1 & a_1 \\
c_2 & a_2
\end{array}\right|, \quad c=\left|\begin{array}{ll}
a_1 & b_1 \\
a_2 & b_2
\end{array}\right| .
$$

***Nota:*** En la práctica, los tres determinantes del teorema anterior pueden obtenerse simplemente escribiendo primero los dos sistemas dados de números directores en tres columnas:

$$
\begin{array}{lll}
a_1 & b_1 & c_1 \\
a_2 & b_2 & c_2
\end{array}
$$

El primer determinante se forma de las segunda y tercera columnas, el segundo de las tercera y primera columas y el tercero de las primera y segunda columnas.

Nos referiremos en adelante a este esquema como el *artificio de los números directores*.

### Ejemplo

Hallar un sistema de números directores para una recta cualquiera $l$ que sea perpendicular al plano que contiene el tríángulo cuyos vértices son $P_1(2,-1,1), P_2(-3,2,2)$ y $P_3(3,3,-2)$.

Por un corolario anterior, dos sistemas de números directores para los lados $P_1 P_2 \quad P_1 P_3$ son, respectivamente.

$$[-3-2,2+1,2-1], \text { o sea, }[-5,3,1]$$

y

$$[3-2,3+1,-2-1], \text { o sea, }[1,4,-3]$$


Por tanto, por el artificio de los números directores, los números directores de $l$ son
$$
\left|\begin{array}{rr}
3 & 1 \\
4 & -3
\end{array}\right|=-13,\left|\begin{array}{rr}
1 & -5 \\
-3 & 1
\end{array}\right|=-14, \quad\left|\begin{array}{rr}
-5 & 3 \\
1 & 4
\end{array}\right|=-23 .
$$

Los resultados de este capítulo son de importancia fundamental en el estudio de la Geometría analítica del espacio. Por esto se recomienda al estudiante que haga un cuadro resumen con todos ellos.
