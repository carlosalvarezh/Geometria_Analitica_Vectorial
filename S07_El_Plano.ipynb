## Forma general de la ecuaci√≥n del plano

La ecuaci√≥n general de un plano en el espacio tridimensional proviene de la necesidad de expresar una relaci√≥n lineal entre las coordenadas $x$, $y$, y $z$ de cualquier punto $P(x, y, z)$ que pertenezca al plano. Tomando como punto de partida la figura proporcionada, consideremos un punto fijo $P_1(x_1, y_1, z_1)$ y una recta normal $\overrightarrow{n}$ con direcci√≥n dada por los n√∫meros directores $[A, B, C]$. 

Para que cualquier otro punto $P(x, y, z)$ est√© en el mismo plano que $P_1$, el vector formado entre $P$ y $P_1$, es decir, $\overrightarrow{P_1P} = [x - x_1, y - y_1, z - z_1]$, debe ser perpendicular a la normal $\overrightarrow{n}$. La condici√≥n de perpendicularidad se satisface mediante el producto punto entre $\overrightarrow{P_1P}$ y el vector normal $[A, B, C]$, el cual debe ser cero:

$$A(x - x_1) + B(y - y_1) + C(z - z_1) = 0.$$

Esta ecuaci√≥n se denomina la ecuaci√≥n escalar del plano, y define la relaci√≥n lineal necesaria entre las coordenadas de cualquier punto que se encuentre en el plano.

Ahora, si reorganizamos los t√©rminos y pasamos $Ax_1 + By_1 + Cz_1$ al otro lado de la ecuaci√≥n, obtenemos la forma general de la ecuaci√≥n del plano:

$$Ax + By + Cz + D = 0,$$

donde $D$ es el negativo del producto punto del vector normal $[A, B, C]$ y el vector de posici√≥n de $P_1$, es decir, $D = -(Ax_1 + By_1 + Cz_1)$. 

En la imagen, se visualiza c√≥mo el plano se extiende infinitamente en todas las direcciones del espacio tridimensional, y c√≥mo cualquier punto en el plano satisface la relaci√≥n expresada por la ecuaci√≥n del plano. De este modo, la ecuaci√≥n generalizada del plano proporciona una herramienta universal para identificar si un punto dado pertenece o no a un plano espec√≠fico en el espacio.

- ***Ejemplo 01:*** Encontrar la ecuaci√≥n del plano que tiene al punto $(3,-2,5)$ y que es normal al vector $(-2,5,3)$.

Para encontrar la ecuaci√≥n de un plano, necesitamos un punto por el cual pase el plano y un vector normal a ese plano. En el ejemplo proporcionado, se nos da el punto $ P_1(3, -2, 5) $ y el vector normal $ n = [-2, 5, 3] $. Utilizando la forma general de la ecuaci√≥n del plano, que es:

$$ Ax + By + Cz + D = 0, $$

podemos identificar $ A $, $ B $, y $ C $ como los componentes del vector normal al plano. En este caso, tendremos:

$$ A = -2, B = 5, C = 3. $$

Para encontrar $ D $, sustituimos el punto $ P_1 $ y los valores de $ A $, $ B $, y $ C $ en la ecuaci√≥n del plano:

$$ -2x + 5y + 3z + D = 0. $$

Sustituyendo $ x = 3 $, $ y = -2 $, y $ z = 5 $ obtenemos:

$$ -2(3) + 5(-2) + 3(5) + D = 0, $$
$$ -6 - 10 + 15 + D = 0, $$
$$ D = 1. $$

Por lo tanto, la ecuaci√≥n del plano buscado es:

$$ -2x + 5y + 3z + 1 = 0. $$

Esta es la ecuaci√≥n del plano que pasa por el punto $ (3, -2, 5) $ y que es perpendicular al vector $ (-2, 5, 3) $.

## Trazas en un plano

Las trazas de un plano son las intersecciones de este con los planos coordenados, y nos permiten tener una visi√≥n m√°s clara de la orientaci√≥n del plano en el espacio tridimensional. Estas trazas se visualizan como rectas en los planos coordenados $ xy $, $ yz $, y $ zx $.

Considerando la ecuaci√≥n general de un plano:

$$Ax + By + Cz + D = 0,$$

las trazas en cada plano coordenado se obtienen al establecer una de las variables $ x $, $ y $, o $ z $ en cero, de la siguiente manera:

- **Plano $ xy $ (cuando $ z = 0 $):** La traza en este plano se obtiene estableciendo $ z = 0 $ en la ecuaci√≥n del plano. Esto nos da la ecuaci√≥n de la recta en el plano $ xy $:

$$Ax + By + D = 0.$$

Esta es la ecuaci√≥n de una recta en el plano $ xy $, que muestra la intersecci√≥n del plano con el plano horizontal $ xy $.

- **Plano $ yz $ (cuando $ x = 0 $):** An√°logamente, establecemos $ x = 0 $ para encontrar la traza en el plano $ yz $:

$$By + Cz + D = 0.$$

Esta es la ecuaci√≥n de una recta en el plano $ yz $ y representa la proyecci√≥n del plano sobre el plano vertical $ yz $.

- **Plano $ zx $ (cuando $ y = 0 $):** Finalmente, al hacer $ y = 0 $, encontramos la traza en el plano $ zx $:

$$Ax + Cz + D = 0.$$

Esta recta en el plano $ zx $ muestra c√≥mo el plano dado corta el plano vertical $ zx $.

Las trazas nos permiten entender la inclinaci√≥n y la orientaci√≥n del plano en el espacio y son muy √∫tiles, por ejemplo, para dibujar el plano en una representaci√≥n tridimensional. Tambi√©n facilitan la comprensi√≥n de c√≥mo un plano puede cortar a trav√©s de las diferentes dimensiones del espacio y ayudan en la soluci√≥n de sistemas de ecuaciones que involucran m√∫ltiples planos.

## Intercepto de un plano con respecto a los ejes

Los interceptos de un plano con respecto a los ejes coordenados son los puntos donde el plano cruza los ejes $ x $, $ y $, y $ z $. Cada uno de estos puntos se encuentra al fijar las otras dos variables en cero y resolviendo para la variable restante. Estos puntos son √∫tiles para esbozar el plano en un sistema de coordenadas y para entender su orientaci√≥n espacial.

Dada la ecuaci√≥n general de un plano:

$$ Ax + By + Cz + D = 0, $$

los interceptos se encuentran de la siguiente manera:

- **Intercepto en el eje $ x $ (cuando $ y = 0 $ y $ z = 0 $):** Al establecer $ y $ y $ z $ en cero, la ecuaci√≥n se simplifica a $ Ax + D = 0 $. Resolviendo para $ x $, obtenemos:

$$ x = -\frac{D}{A}. $$

Si $ A $ es cero, el plano es paralelo al eje $ x $ y no hay intercepto con el eje $ x $.

- **Intercepto en el eje $ y $ (cuando $ x = 0 $ y $ z = 0 $):** De manera similar, si $ x $ y $ z $ son cero, tenemos $ By + D = 0 $. Resolviendo para $ y $, encontramos:

$$ y = -\frac{D}{B}. $$

Si $ B $ es cero, el plano es paralelo al eje $ y $ y no hay intercepto con el eje $ y $.

- **Intercepto en el eje $ z $ (cuando $ x = 0 $ y $ y = 0 $):** Por √∫ltimo, con $ x $ y $ y $ en cero, la ecuaci√≥n se convierte en $ Cz + D = 0 $. Resolviendo para $ z $, se obtiene:

$$ z = -\frac{D}{C}. $$

Si $ C $ es cero, el plano es paralelo al eje $ z $ y no hay intercepto con el eje $ z $.

Los interceptos son com√∫nmente representados por los puntos $ (a, 0, 0) $, $ (0, b, 0) $, y $ (0, 0, c) $, donde $ a = -\frac{D}{A} $, $ b = -\frac{D}{B} $, y $ c = -\frac{D}{C} $. Estos valores representan las distancias desde el origen a los puntos donde el plano intercepta los ejes $ x $, $ y $, y $ z $, respectivamente. Si uno de los denominadores es cero, eso indica que el plano es paralelo al eje correspondiente y por lo tanto no hay intercepto real. Si $ D $ es cero, entonces uno o m√°s de los interceptos estar√°n en el origen, indicando que el plano pasa por el origen del sistema de coordenadas.

## Forma sim√©trica del plano

La forma sim√©trica de la ecuaci√≥n de un plano es una representaci√≥n que relaciona cada coordenada del punto $ (x, y, z) $ con sus interceptos respectivos sobre los ejes $ x $, $ y $, y $ z $. Es √∫til para identificar r√°pidamente estos interceptos y para entender c√≥mo var√≠an las coordenadas del punto a lo largo del plano.

Partiendo de la ecuaci√≥n general del plano:

$$
Ax + By + Cz + D = 0,
$$

y dividiendo cada t√©rmino por $ -D $ (asumiendo que $ D \neq 0 $), obtenemos:

$$
\frac{Ax}{-D} + \frac{By}{-D} + \frac{Cz}{-D} - 1 = 0.
$$

Reescribiendo esta ecuaci√≥n en t√©rminos de los interceptos $ a = -\frac{D}{A} $, $ b = -\frac{D}{B} $, y $ c = -\frac{D}{C} $, tenemos:

$$
\frac{x}{a} + \frac{y}{b} + \frac{z}{c} - 1 = 0,
$$

lo que simplificado da la forma sim√©trica de la ecuaci√≥n del plano:

$$
\frac{x}{a} + \frac{y}{b} + \frac{z}{c} = 1.
$$

Esta forma sim√©trica nos dice que para cualquier punto $ (x, y, z) $ que est√© en el plano, la suma de las fracciones de sus coordenadas divididas por sus interceptos correspondientes es igual a uno. Es importante destacar que esta forma solo es v√°lida cuando $ A $, $ B $, y $ C $ no son cero, ya que de lo contrario los interceptos no estar√≠an definidos (el plano ser√≠a paralelo a uno de los ejes y no habr√≠a intersecci√≥n).

La forma sim√©trica de la ecuaci√≥n del plano es particularmente √∫til cuando se trabaja con planos en problemas de geometr√≠a descriptiva y cuando se requiere una r√°pida visualizaci√≥n de las proporciones del plano con respecto a los ejes del sistema de coordenadas. Adem√°s, proporciona una manera elegante de confirmar si un punto dado est√° sobre el plano simplemente verificando si la suma de las coordenadas del punto, cada una dividida por su intercepto correspondiente, es igual a uno.

- ***Ejemplo 04:*** Para el siguiente plano encontrar los puntos de intersecci√≥n con los ejes coordenados y las ecuaciones de sus trazas sobre los planos coordenados, adem√°s encuentre la forma sim√©trica de su ecuaci√≥n $m_1: 4 x+6 y+3 z-12=0$.

Para resolver el ejemplo proporcionado, seguiremos estos pasos:

$$4x + 6y + 3z - 12 = 0$$

Punto de intersecci√≥n: $ (0, 2, 0) $.

- **Intersecci√≥n con el eje $ z $ (cuando $ x = 0 $ y $ y = 0 $):**
$$3z - 12 = 0 \implies z = \frac{12}{3} = 4.$$
Punto de intersecci√≥n: $ (0, 0, 4) $.


2. **Encontrar las ecuaciones de las trazas sobre los planos coordenados**

- **Plano $ xy $ (cuando $ z = 0 $):**
$$4x + 6y - 12 = 0$$

- **Plano $ yz $ (cuando $ x = 0 $):**
$$6y + 3z - 12 = 0$$

- **Plano $ zx $ (cuando $ y = 0 $):**

Estos resultados proporcionan una descripci√≥n completa del plano en t√©rminos de sus intersecciones con los ejes, las trazas en los planos coordenados y su representaci√≥n en forma sim√©trica, facilitando su visualizaci√≥n y an√°lisis en problemas geom√©tricos y de dise√±o.

- ***Ejemplo 06.*** Hallar la ecuaci√≥n del plano perpendicular al plano $X Y$ y que pasa por los puntos $P_1(1,5,-3)$ y $P_2(-5,-4,11)$.

Para encontrar la ecuaci√≥n de un plano que es perpendicular al plano $XY$ y que pasa por dos puntos dados, debemos entender que un plano perpendicular al plano $XY$ es paralelo al eje $z$, y su ecuaci√≥n solo involucra las variables $x$ y $y$, ya que la componente $z$ es irrelevante (no influye en la orientaci√≥n del plano en relaci√≥n con el eje $z$). $$

Usamos los puntos $P_1$ y $P_2$ para determinar los coeficientes $A$ y $B$, sabiendo que ambos deben satisfacer la ecuaci√≥n del plano.

**Formulamos un sistema de ecuaciones basado en estos puntos:**
1. Sustituyendo $P_1(1, 5)$:
   $$ A(1) + B(5) + D = 0 $$
   $$ A + 5B + D = 0 $$

2. Sustituyendo $P_2(-5, -4)$:
   $$ A(-5) + B(-4) + D = 0 $$
   $$ -5A - 4B + D = 0 $$

**Resolver este sistema de ecuaciones lineales:**

## √Ångulo entre dos planos

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Plano02.PNG?raw=true" width="350" />
</p>

El √°ngulo entre dos planos en el espacio tridimensional puede ser entendido como el √°ngulo entre sus vectores normales. En la geometr√≠a anal√≠tica, este √°ngulo se puede calcular utilizando el producto escalar (tambi√©n conocido como producto punto) de los vectores normales de los dos planos.

El producto escalar de dos vectores $\overrightarrow{a}$ y $\overrightarrow{b}$ se define como:

$$
\overrightarrow{a} \cdot \overrightarrow{b} = |\overrightarrow{a}| |\overrightarrow{b}| \cos(\theta)
$$

donde:
- $|\overrightarrow{a}|$ es la magnitud o longitud del vector $\overrightarrow{a}$,
- $|\overrightarrow{b}|$ es la magnitud o longitud del vector $\overrightarrow{b}$,
- $\theta$ es el √°ngulo entre los vectores $\overrightarrow{a}$ y $\overrightarrow{b}$.

Para calcular el √°ngulo $\theta$ entre dos vectores normales $\overrightarrow{n_1}$ y $\overrightarrow{n_2}$, reorganizamos la f√≥rmula para resolver para $\cos(\theta)$:

$$
\cos(\theta) = \frac{\overrightarrow{n_1} \cdot \overrightarrow{n_2}}{|\overrightarrow{n_1}| |\overrightarrow{n_2}|}
$$

El producto escalar $\overrightarrow{n_1} \cdot \overrightarrow{n_2}$ se calcula como:

$$
\overrightarrow{n_1} \cdot \overrightarrow{n_2} = n_{1x}n_{2x} + n_{1y}n_{2y} + n_{1z}n_{2z}
$$

Y las magnitudes de los vectores normales se calculan como:

$$
|\overrightarrow{n_1}| = \sqrt{n_{1x}^2 + n_{1y}^2 + n_{1z}^2}
$$
$$
|\overrightarrow{n_2}| = \sqrt{n_{2x}^2 + n_{2y}^2 + n_{2z}^2}
$$

Una vez que tenemos $\cos(\theta)$, podemos encontrar el √°ngulo $\theta$ usando la funci√≥n arccos (la funci√≥n inversa del coseno):

$$
\theta = \arccos\left(\frac{\overrightarrow{n_1} \cdot \overrightarrow{n_2}}{|\overrightarrow{n_1}| |\overrightarrow{n_2}|}\right)
$$

Es importante notar que el √°ngulo resultante siempre ser√° el menor de los dos √°ngulos posibles entre los planos y estar√° en el rango de $0^\circ$ a $90^\circ$. Si se desea el √°ngulo obtuso, se puede restar el √°ngulo calculado de $180^\circ$.

Adem√°s, cuando los planos son perpendiculares, el producto escalar ser√° cero, y cuando son paralelos, el coseno ser√° $\pm 1$, lo que indica un √°ngulo de $0^\circ$ o $180^\circ$ entre ellos.

- ***Ejemplo08:*** Encontrar el √°ngulo entre los planos $m_1: 3x + 2y + 8z - 9 = 0$ y $m_2: 5x + 6y - 2z + 3 = 0$

Para encontrar el √°ngulo entre los dos planos dados, utilizamos sus vectores normales. Los vectores normales pueden obtenerse directamente de los coeficientes de las variables $x$, $y$, y $z$ en las ecuaciones de los planos.

Para el plano $m_1: 3x + 2y + 8z - 9 = 0$, el vector normal es $\vec{n_1} = [3, 2, 8]$.

Para el plano $m_2: 5x + 6y - 2z + 3 = 0$, el vector normal es $\vec{n_2} = [5, 6, -2]$.

El √°ngulo entre los dos vectores normales, y por lo tanto entre los planos, se encuentra con la f√≥rmula del producto escalar:

$$\cos(\theta) = \frac{\vec{n_1} \cdot \vec{n_2}}{|\vec{n_1}| |\vec{n_2}|}$$

El producto escalar $\vec{n_1} \cdot \vec{n_2}$ se calcula como:

$$\vec{n_1} \cdot \vec{n_2} = (3)(5) + (2)(6) + (8)(-2)$$

Las magnitudes de $\vec{n_1}$ y $\vec{n_2}$ se calculan como:

$$|\vec{n_1}| = \sqrt{3^2 + 2^2 + 8^2}$$
$$|\vec{n_2}| = \sqrt{5^2 + 6^2 + (-2)^2}$$

Podemos calcular estos valores para obtener $\cos(\theta)$ y luego usar la funci√≥n arccos para encontrar $\theta$. Realicemos los c√°lculos.

Parece que hay un problema t√©cnico que me impide realizar los c√°lculos en este momento. A pesar de eso, te puedo guiar sobre c√≥mo completar los c√°lculos.

Usando las f√≥rmulas que proporcion√©, puedes calcular el producto escalar (dot product) y las magnitudes (norms) de los vectores normales de los planos de la siguiente manera:

**Producto escalar:**
$$
\vec{n_1} \cdot \vec{n_2} = (3 \cdot 5) + (2 \cdot 6) + (8 \cdot -2) = 15 + 12 - 16 = 11
$$

**Magnitudes:**
$$
|\vec{n_1}| = \sqrt{3^2 + 2^2 + 8^2} = \sqrt{9 + 4 + 64} = \sqrt{77}
$$
$$
|\vec{n_2}| = \sqrt{5^2 + 6^2 + (-2)^2} = \sqrt{25 + 36 + 4} = \sqrt{65}
$$

**Coseno del √°ngulo:**
$$
\cos(\theta) = \frac{11}{\sqrt{77} \cdot \sqrt{65}}
$$

Con esos valores, puedes usar una calculadora para encontrar $\cos(\theta)$ y luego calcular $\theta$ en grados:

$$
\theta = \arccos\left(\frac{11}{\sqrt{77} \cdot \sqrt{65}}\right)
$$

Aseg√∫rate de tener tu calculadora en el modo adecuado (radianes o grados) dependiendo de c√≥mo desees el resultado. Si deseas realizar los c√°lculos en una calculadora o software que tengas a mano, deber√≠as obtener la medida del √°ngulo entre los planos $m_1$ y $m_2$.

```python
import numpy as np

# Definir los vectores normales de los dos planos
n1 = np.array([3, 2, 8])
n2 = np.array([5, 6, -2])

# Calcular el producto escalar (punto) de n1 y n2
dot_product = np.dot(n1, n2)

# Calcular las magnitudes de n1 y n2
magnitude_n1 = np.linalg.norm(n1)
magnitude_n2 = np.linalg.norm(n2)

# Calcular el coseno del √°ngulo theta
cos_theta = dot_product / (magnitude_n1 * magnitude_n2)

# Calcular el √°ngulo theta en grados
theta_radians = np.arccos(cos_theta)
theta_degrees = np.degrees(theta_radians)

cos_theta, theta_degrees
```

## Intersecci√≥n entre dos planos

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Plano03.PNG?raw=true" width="250" />
</p>

La intersecci√≥n de dos planos no paralelos en el espacio tridimensional es una recta. Cuando dos planos se cortan, sus ecuaciones pueden usarse para definir un√≠vocamente la recta de intersecci√≥n. Las ecuaciones de los planos en su forma general son:

$$
\begin{aligned}
& m_1: A_1 x + B_1 y + C_1 z + D_1 = 0 \\
& m_2: A_2 x + B_2 y + C_2 z + D_2 = 0
\end{aligned}
$$

Para que dos planos se intersecten, deben cumplir con las siguientes condiciones:

1. **No son paralelos**: Esto significa que sus vectores normales no son m√∫ltiplos escalares el uno del otro. Matem√°ticamente, esto se expresa como $ A_1/B_1 \neq A_2/B_2 $ o $ B_1/C_1 \neq B_2/C_2 $, lo que implica que el √°ngulo $ \theta $ entre los vectores normales $ \vec{n}_1 $ y $ \vec{n}_2 $ no es cero (o sea, $ \theta \neq 0 $).

2. **No son coincidentes**: Esto significa que no tienen exactamente las mismas ecuaciones o no son m√∫ltiplos escalares uno del otro.

Cuando se resuelve el sistema formado por las ecuaciones de los dos planos, el resultado es una serie de soluciones que representan los puntos a trav√©s de los cuales pasa la recta de intersecci√≥n. Este sistema generalmente es compatible e indeterminado, ya que hay infinitos puntos a lo largo de la recta que satisfacen ambas ecuaciones.

Para expresar la recta de intersecci√≥n de forma param√©trica o sim√©trica, se puede hacer lo siguiente:

- Seleccionar una variable para ser el par√°metro, com√∫nmente $ t $, y expresar las otras dos en funci√≥n de este par√°metro.
- Resolver el sistema de ecuaciones de los dos planos para dos de las variables en t√©rminos de la tercera.
- Si es posible, eliminar el par√°metro para obtener la ecuaci√≥n sim√©trica de la recta.

La ecuaci√≥n param√©trica de la recta tiene la forma:
$$
\begin{aligned}
& x = x_0 + at \\
& y = y_0 + bt \\
& z = z_0 + ct
\end{aligned}
$$
donde $ (x_0, y_0, z_0) $ es un punto por el que pasa la recta y $ (a, b, c) $ es la direcci√≥n de la recta, la cual se puede obtener tambi√©n como el producto cruz de los vectores normales de los planos.

La ecuaci√≥n sim√©trica se deriva de la param√©trica dividiendo cada una de las ecuaciones param√©tricas por sus coeficientes direccionales correspondientes, asumiendo que son distintos de cero.

Conocer la recta de intersecci√≥n entre dos planos es √∫til en diversas aplicaciones, como la construcci√≥n de objetos tridimensionales en dise√±o asistido por computadora, la soluci√≥n de problemas en la geometr√≠a del espacio y en la f√≠sica, especialmente en mec√°nica y √≥ptica, donde las l√≠neas de acci√≥n y las trayectorias suelen ser rectas que resultan de la intersecci√≥n de planos de fuerzas o superficies reflectantes/refractantes.

- ***Ejemplo 09:*** Obtener la ecuaci√≥n de la recta producto de la  intersecci√≥n de los planos $ùëö1 : ùë• ‚àí ùë¶ ‚àí ùëß = 0$ y $ùëö2 : ùë• ‚àí 2ùë¶ + ùëß ‚àí 1 = 0$.

Para encontrar la ecuaci√≥n de la recta que es la intersecci√≥n de los dos planos dados, $m_1: x - y - z = 0$ y $m_2: x - 2y + z - 1 = 0$, necesitamos resolver el sistema de ecuaciones que estos forman. En este caso, como estamos en tres dimensiones y solo tenemos dos ecuaciones, esperamos que la soluci√≥n del sistema nos d√© una recta.

Un m√©todo para resolver este sistema es expresar dos de las variables en t√©rminos de la tercera y luego usar esta tercera variable como un par√°metro libre para las ecuaciones param√©tricas de la recta. Veamos c√≥mo hacerlo paso a paso.

***Paso 1: Reescribir las ecuaciones:*** Reescribimos las ecuaciones de los planos

$$ m_1: x - y - z = 0$$
$$ m_2: x - 2y + z - 1 = 0$$

***Paso 2: Resolver el sistema de ecuaciones*** Podemos sumar $ m_1 $ y $ m_2 $ para eliminar la variable $ z $. Sumando $ m_1 $ y $ m_2 $:
   $$ (x - y - z) + (x - 2y + z - 1) = 0 $$
   $$ 2x - 3y - 1 = 0 $$

Podemos resolver esta ecuaci√≥n para $ x $ o $ y $ en t√©rminos del otro. Escojamos $ y $ como nuestro par√°metro libre y expresemos $ x $ en t√©rminos de $ y $:

$$ 2x = 3y + 1 $$
$$ x = \frac{3}{2}y + \frac{1}{2} $$

Ahora, sustituimos $ x $ en $ m_1 $ para encontrar $ z $ en t√©rminos de $ y $:

$$ \frac{3}{2}y + \frac{1}{2} - y - z = 0 $$
$$ \frac{1}{2}y + \frac{1}{2} - z = 0 $$
$$ z = \frac{1}{2}y + \frac{1}{2} $$

***Paso 3: Ecuaciones param√©tricas de la recta:*** Ahora que hemos expresado $ x $ y $ z $ en t√©rminos de $ y $, podemos escribir las ecuaciones param√©tricas de la recta usando $ y $ como el par√°metro (denotado como $ t $):

$$ x = \frac{3}{2}t + \frac{1}{2} $$
$$ y = t $$
$$ z = \frac{1}{2}t + \frac{1}{2} $$

***Resultado: Ecuaci√≥n param√©trica de la recta de intersecci√≥n:*** La ecuaci√≥n param√©trica de la recta de intersecci√≥n es entonces

$$
\begin{aligned}
x &= \frac{3}{2}t + \frac{1}{2} \\
y &= t \\
z &= \frac{1}{2}t + \frac{1}{2}
\end{aligned}
$$

Donde $ t $ es el par√°metro que recorre la recta. Estas ecuaciones describen la serie de puntos por los que pasa la recta que es la intersecci√≥n de los dos planos dados.

## Distancia de un punto a un plano

La distancia m√°s corta de un punto a un plano es perpendicular al plano y se puede calcular utilizando el vector normal al plano y la posici√≥n del punto en cuesti√≥n. 

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Plano04.PNG?raw=true" width="350" />
</p>

Consideremos un plano en el espacio definido por la ecuaci√≥n:

$$ m: A x + B y + C z + D = 0 $$

donde $ A $, $ B $, y $ C $ son los coeficientes que corresponden al vector normal del plano, $ \vec{n} $, y $ D $ es el t√©rmino constante. El vector normal es una componente esencial, ya que dicta la orientaci√≥n del plano en el espacio y es perpendicular a cualquier vector que yace en el plano.

Ahora, sea $ P(x_1, y_1, z_1) $ un punto fuera del plano. La distancia $ d $ entre el punto $ P $ y el plano $ m $ es el largo del segmento perpendicular desde $ P $ hasta $ m $. Este segmento representa la hipotenusa en un tri√°ngulo rect√°ngulo imaginario cuyos catetos yacen en el plano y en la l√≠nea perpendicular que pasa por $ P $.


1. **Vector normal $ \vec{n} $**: Este vector se obtiene directamente de los coeficientes de la ecuaci√≥n del plano y est√° representado como $ \vec{n} = [A, B, C] $.

2. **Vector unitario $ \hat{n} $**: El vector normal unitario se calcula normalizando el vector normal $ \vec{n} $, dividi√©ndolo entre su magnitud. Es decir:
   $$
   \hat{n} = \frac{\vec{n}}{|\vec{n}|}
   $$

3. **Magnitud del vector normal**: La magnitud $ |\vec{n}| $ es la ra√≠z cuadrada de la suma de los cuadrados de los componentes del vector normal, que es:
   $$
   |\vec{n}| = \sqrt{A^2 + B^2 + C^2}
   $$

4. **Vector de posici√≥n $ \vec{P} $**: Es el vector desde un punto conocido $ x_0, y_0, z_0 $ en el plano (usualmente se toma como el punto m√°s cercano en el plano al origen para simplificar los c√°lculos) hasta el punto $ P(x_1, y_1, z_1) $ cuya distancia queremos calcular, y est√° dado por:

$$\vec{P}=\left(x_1-x_0\right) \hat{i} + \left(y_1-y_0\right) \hat{j} + \left(z_1-x_z\right) \hat{k}$$

5. **Proyecci√≥n de $ \vec{P} $ sobre $ \hat{n} $**: La distancia m√°s corta entre el plano y el punto $ P $ es la longitud de la proyecci√≥n del vector $ \vec{P} $ sobre el vector unitario normal $ \hat{n} $, y se calcula como:
   $$
   d = |\vec{P} \cdot \hat{n}|
   $$

6. **Ecuaci√≥n de la distancia**: La ecuaci√≥n de la distancia se simplifica al tomar en cuenta que el punto $ x_0, y_0, z_0 $ que se utiliza para definir el vector de posici√≥n $ \vec{P} $ es en realidad un punto del plano, por lo que satisface la ecuaci√≥n del plano $ Ax_0 + By_0 + Cz_0 + D = 0 $. Por lo tanto, la ecuaci√≥n final de la distancia $ d $ del punto $ P $ al plano $ m $ es:
   $$
   d = \left|\frac{A x_1 + B y_1 + C z_1 + D}{\sqrt{A^2 + B^2 + C^2}}\right|
   $$

Donde $ A, B, C $ son los coeficientes de la ecuaci√≥n del plano y $ D $ es el t√©rmino constante. La distancia $ d $ se obtiene tomando el valor absoluto de esta expresi√≥n, ya que la distancia es siempre un valor positivo. Este resultado se basa en la f√≥rmula del producto punto y la proyecci√≥n de un vector sobre otro, y es una aplicaci√≥n directa del teorema de Pit√°goras en el espacio tridimensional.

Distancia de un punto a un plano
Para un plano $m$ conocido $\left(\begin{array}{lll}\vec{n} & y & x_0\end{array}\right)$ y un punto $P\left(x_1, y_1, z_1\right)$ dado.
La distancia $d$ mas corta entre el plano y el punto $P$ es una distancia paralela al vector normal $\vec{n}$ al plano $m$.
Donde esta distancia $d$ se obtiene a partir de la proyecci√≥n del vector $\vec{P}$ en $d=|\vec{P} \cdot \hat{n}|$ el vector unitario $\vec{n}$.
donde:
$$
\begin{gathered}
\hat{n}=\frac{\vec{n}}{|\vec{n}|} \Rightarrow \hat{n}=\frac{n_x \hat{\imath}+n_y \hat{\jmath}+n_z \hat{\mathrm{k}}}{\sqrt{n_x^2+n_y^2+n_z^2}} \\
\vec{P}=\left(x_1-x_0\right) \hat{\imath}+\left(y_1-y_0\right) \hat{\jmath}+\left(z_1-z_0\right) \hat{\mathrm{k}}, \text { sustituyendo } \\
d=\left|\frac{n_x\left(x_1-x_0\right)+n_y\left(y_1-y_0\right)+n_z\left(z_1-z_0\right)}{\sqrt{n_x^2+n_y^2+n_z^2}}\right| \\
d=\left|\frac{A x_1+B y_1+C z_1+D}{\sqrt{A^2+B^2+C^2}}\right|
\end{gathered}
$$

Una familia de planos que pasa por un punto espec√≠fico comparte ese punto en com√∫n. El punto puede considerarse un v√©rtice o un centro desde el cual "irradian" los diferentes planos.

Para un punto dado $P(x_1, y_1, z_1)$, la ecuaci√≥n del plano que pasa por este punto puede escribirse de la siguiente manera:

$$A(x - x_1) + B(y - y_1) + C(z - z_1) = 0$$

Aqu√≠, $A$, $B$, y $C$ representan los coeficientes del vector normal al plano, $\vec{n} = (A, B, C)$, que dicta la orientaci√≥n del plano. La ecuaci√≥n tambi√©n se puede expandir para obtener la forma est√°ndar:

$$Ax + By + Cz = Ax_1 + By_1 + Cz_1$$

El punto de intersecci√≥n, $P(x_1, y_1, z_1)$, se denomina punto o v√©rtice de radiaci√≥n, ya que es el punto donde todas las superficies de la familia convergen. Cambiando los valores de $A$, $B$, y $C$, se pueden generar infinitos planos que pasan por $P$ con diferentes inclinaciones.

### Familia de Planos Paralelos

Otra familia com√∫n es la de los planos paralelos, que mantienen una distancia constante entre s√≠. Tienen el mismo vector normal pero difieren en el t√©rmino constante, lo que significa que sus ecuaciones son similares salvo por el valor de $ D $. La ecuaci√≥n de un plano general puede expresarse como:

$$ Ax + By + Cz + D = 0 $$

Donde el vector normal, $ \vec{n} = (A, B, C) $, es el mismo para cada plano de la familia. Solo el valor de $ D $ cambia, lo que permite generar m√∫ltiples planos con la misma orientaci√≥n pero diferentes posiciones.

### Familia de Planos Perpendiculares a una Recta

Los planos perpendiculares a una recta determinada forman otra familia interesante. Si una recta dada tiene direcci√≥n $ \vec{d} = (a, b, c) $, cualquier plano cuyo vector normal sea paralelo o antiparalelo a $ \vec{d} $ ser√° perpendicular a la recta.

La ecuaci√≥n del plano perpendicular a la recta puede escribirse en la forma:

$$ a(x - x_0) + b(y - y_0) + c(z - z_0) = 0 $$

donde $ (x_0, y_0, z_0) $ es un punto en el plano.

### Familia de Planos que Contienen la Recta Generada por la Intersecci√≥n de dos Planos

Cuando dos planos en el espacio tridimensional se intersectan, lo hacen en una l√≠nea recta. Esta l√≠nea representa el conjunto de puntos que satisfacen simult√°neamente las ecuaciones de ambos planos. La familia de planos que pasan por esta l√≠nea de intersecci√≥n pueden describirse como una combinaci√≥n lineal de las ecuaciones de los dos planos originales.

Dado el conjunto de ecuaciones:

$$ m_1: A_1 x + B_1 y + C_1 z + D_1 = 0 $$
$$ m_2: A_2 x + B_2 y + C_2 z + D_2 = 0 $$

la ecuaci√≥n de cualquier plano que contiene la l√≠nea de intersecci√≥n de los planos $ m_1 $ y $ m_2 $ se expresa como:

$$ A_1 x + B_1 y + C_1 z + D_1 + k(A_2 x + B_2 y + C_2 z + D_2) = 0 $$

donde $ k $ es un par√°metro arbitrario que ajusta el peso de las contribuciones de cada plano. Al expandir esta ecuaci√≥n, se observa que todos los planos de esta familia comparten la recta de intersecci√≥n, ya que satisfacen ambas ecuaciones simult√°neamente.

***Propiedades***

1. **Estructura General:** La ecuaci√≥n representa la combinaci√≥n de las ecuaciones de ambos planos, y el par√°metro $ k $ proporciona la flexibilidad para generar infinitos planos que pasan por la misma l√≠nea de intersecci√≥n.

2. **Vector Normal:** El vector normal de cualquiera de estos planos es una combinaci√≥n lineal de los vectores normales de los dos planos originales.

3. **Aplicaci√≥n:** Esta familia de planos puede utilizarse para identificar superficies que comparten un borde com√∫n, dise√±ar estructuras espaciales que pasan por una misma l√≠nea de referencia, o crear modelos que requieran superficies alineadas.


- ***Ejemplo 11:*** Encontrar la propiedad en com√∫n de los siguientes planos:  
***a)***
$$
\begin{aligned}
& m_1: 9 x+4 y-z=0 \\
& m_2: 9 x+4 y-z=0
\end{aligned}
$$
***b)***
$$
\begin{aligned}
& m_1: x-5 y+2 z-4=0 \\
& m_2: 2 x-10 y+4 z-6=0
\end{aligned}
$$
***c)***
$$
\begin{aligned}
& m_1: 5(x-2)+7(y-7)-\frac{5}{2}(z+2)=0 \\
& m_2: 3(x-2)-4(y-7)+9(z+2)=0
\end{aligned}
$$

***Soluci√≥n***

Para identificar la propiedad com√∫n en los ejemplos de planos proporcionados, vamos a analizar cada par de ecuaciones para entender la relaci√≥n entre ellos.

***a)***
$$ m_1: 9x + 4y - z = 0 $$
$$ m_2: 9x + 4y - z = 0 $$

Los dos planos $ m_1 $ y $ m_2 $ tienen exactamente la misma ecuaci√≥n, lo que significa que son el mismo plano. Por lo tanto, comparten todas sus propiedades geom√©tricas, incluyendo su vector normal, su posici√≥n y su orientaci√≥n.

**Propiedad en com√∫n:** Estos dos planos son id√©nticos.

***b)***
$$ m_1: x - 5y + 2z - 4 = 0 $$
$$ m_2: 2x - 10y + 4z - 6 = 0 $$

Para analizar la relaci√≥n entre estos dos planos, podemos notar que $ m_2 $ es un m√∫ltiplo escalar de $ m_1 $. Al multiplicar la ecuaci√≥n de $ m_1 $ por 2, obtenemos la ecuaci√≥n de $ m_2 $, excepto por la constante diferente (-6). Esto implica que ambos planos son paralelos, pero est√°n en diferentes posiciones.

**Propiedad en com√∫n:** Estos planos son paralelos entre s√≠.

***c)***
$$ m_1: 5(x - 2) + 7(y - 7) - \frac{5}{2}(z + 2) = 0 $$
$$ m_2: 3(x - 2) - 4(y - 7) + 9(z + 2) = 0 $$

Primero, expandamos las ecuaciones para verlas en su forma est√°ndar.

Para $ m_1 $:
$$ 5x - 10 + 7y - 49 - \frac{5}{2}z - 5 = 0 $$
$$ 5x + 7y - \frac{5}{2}z - 64 = 0 $$

Para $ m_2 $:
$$ 3x - 6 - 4y + 28 + 9z + 18 = 0 $$
$$ 3x - 4y + 9z + 40 = 0 $$

Notamos que ambos planos pasan por el mismo punto $ (2, 7, -2) $. Si sustituimos las coordenadas de este punto en ambas ecuaciones, ambas se satisfacen.

**Propiedad en com√∫n:** Estos planos comparten un punto en com√∫n: $ (2, 7, -2) $.

- ***Ejemplo 12:*** Encontrar la ecuaci√≥n de plano $m_o$ que pasa por la intersecci√≥n de los planos $\quad m_1: x+y+z+3=0 \quad$ y $\quad m_2: 2 x+3 y-5 z=0$ que tiene una distancia con respecto al origen de $\sqrt{3}$.

Para encontrar la ecuaci√≥n del plano $ m_o $ que pasa por la intersecci√≥n de los planos $ m_1 $ y $ m_2 $ y tiene una distancia de $\sqrt{3}$ con respecto al origen, seguiremos estos pasos:

***Paso 1: Escribir la ecuaci√≥n del plano $ m_o $ -*** La ecuaci√≥n del plano $ m_o $ que pasa por la intersecci√≥n de los planos $ m_1 $ y $ m_2 $ puede representarse como una combinaci√≥n lineal de las ecuaciones de ambos planos:
$$ m_o: m_1 + k m_2 = x + y + z + 3 + k (2x + 3y - 5z) = 0 $$
Expandamos esta expresi√≥n para obtener la forma est√°ndar:
$$ x + y + z + 3 + 2kx + 3ky - 5kz = 0 $$
$$ (1 + 2k)x + (1 + 3k)y + (1 - 5k)z + 3 = 0 $$

Esta es la ecuaci√≥n general del plano $ m_o $. Necesitamos encontrar el valor de $ k $ que haga que el plano est√© a una distancia de $\sqrt{3}$ del origen.

***Paso 2: Calcular el vector normal de $ m_o $ -*** El vector normal del plano $ m_o $ es:
$$ \vec{n} = (1 + 2k, 1 + 3k, 1 - 5k) $$

***Paso 3: Usar la f√≥rmula de distancia -*** La distancia entre el plano $ m_o $ y el origen es dada por la f√≥rmula:

$$ d = \frac{|A \cdot 0 + B \cdot 0 + C \cdot 0 + D|}{\sqrt{A^2 + B^2 + C^2}} = \frac{|3|}{\sqrt{(1 + 2k)^2 + (1 + 3k)^2 + (1 - 5k)^2}} $$

Queremos que la distancia sea igual a $\sqrt{3}$:
$$ \sqrt{3} = \frac{|3|}{\sqrt{(1 + 2k)^2 + (1 + 3k)^2 + (1 - 5k)^2}} $$

***Paso 4: Resolver para $ k $ -*** Despejemos el valor de $ k $ que satisface esta ecuaci√≥n y obtendremos la ecuaci√≥n final del plano $ m_o $.

1. **Expresi√≥n del vector normal:**
   El vector normal del plano $ m_o $ es $ \vec{n} = (1 + 2k, 1 + 3k, 1 - 5k) $.

2. **Magnitud del vector normal:**
   La magnitud del vector normal es:
   $$
   |\vec{n}| = \sqrt{(1 + 2k)^2 + (1 + 3k)^2 + (1 - 5k)^2}
   $$

3. **Ecuaci√≥n de distancia:**
   La distancia $ d $ entre el plano y el origen es:
   $$
   d = \frac{|3|}{\sqrt{(1 + 2k)^2 + (1 + 3k)^2 + (1 - 5k)^2}}
   $$
   Queremos que esta distancia sea $\sqrt{3}$:
   $$
   \sqrt{3} = \frac{|3|}{\sqrt{(1 + 2k)^2 + (1 + 3k)^2 + (1 - 5k)^2}}
   $$

4. **Resolver para $ k $:**
   Despejando el valor de $ k $, se obtiene un sistema cuadr√°tico que puede resolverse para encontrar el valor de $ k $. Una vez que tengamos el valor de $ k $, lo sustituimos en la ecuaci√≥n general:
   $$
   (1 + 2k)x + (1 + 3k)y + (1 - 5k)z + 3 = 0
   $$
   y obtenemos la ecuaci√≥n del plano $ m_o $ que tiene una distancia de $\sqrt{3}$ al origen.
