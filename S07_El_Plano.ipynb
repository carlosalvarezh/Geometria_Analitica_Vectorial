## Forma general de la ecuación del plano

  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Plano00.PNG?raw=true" width="250" />
</p>

La ecuación general de un plano en el espacio tridimensional proviene de la necesidad de expresar una relación lineal entre las coordenadas $x$, $y$, y $z$ de cualquier punto $P(x, y, z)$ que pertenezca al plano. Tomando como punto de partida la figura proporcionada, consideremos un punto fijo $P_1(x_1, y_1, z_1)$ y una recta normal $\overrightarrow{n}$ con dirección dada por los números directores $[A, B, C]$. 

Para que cualquier otro punto $P(x, y, z)$ esté en el mismo plano que $P_1$, el vector formado entre $P$ y $P_1$, es decir, $\overrightarrow{P_1P} = [x - x_1, y - y_1, z - z_1]$, debe ser perpendicular a la normal $\overrightarrow{n}$. La condición de perpendicularidad se satisface mediante el producto punto entre $\overrightarrow{P_1P}$ y el vector normal $[A, B, C]$, el cual debe ser cero:

$$A(x - x_1) + B(y - y_1) + C(z - z_1) = 0.$$

Esta ecuación se denomina la ecuación escalar del plano, y define la relación lineal necesaria entre las coordenadas de cualquier punto que se encuentre en el plano.

Ahora, si reorganizamos los términos y pasamos $Ax_1 + By_1 + Cz_1$ al otro lado de la ecuación, obtenemos la forma general de la ecuación del plano:

$$Ax + By + Cz + D = 0,$$

donde $D$ es el negativo del producto punto del vector normal $[A, B, C]$ y el vector de posición de $P_1$, es decir, $D = -(Ax_1 + By_1 + Cz_1)$. 

En la imagen, se visualiza cómo el plano se extiende infinitamente en todas las direcciones del espacio tridimensional, y cómo cualquier punto en el plano satisface la relación expresada por la ecuación del plano. De este modo, la ecuación generalizada del plano proporciona una herramienta universal para identificar si un punto dado pertenece o no a un plano específico en el espacio.

- ***Ejemplo 01:*** Encontrar la ecuación del plano que tiene al punto $(3,-2,5)$ y que es normal al vector $(-2,5,3)$.

Para encontrar la ecuación de un plano, necesitamos un punto por el cual pase el plano y un vector normal a ese plano. En el ejemplo proporcionado, se nos da el punto $ P_1(3, -2, 5) $ y el vector normal $ n = [-2, 5, 3] $. Utilizando la forma general de la ecuación del plano, que es:

$$ Ax + By + Cz + D = 0, $$

podemos identificar $ A $, $ B $, y $ C $ como los componentes del vector normal al plano. En este caso, tendremos:

$$ A = -2, B = 5, C = 3. $$

Para encontrar $ D $, sustituimos el punto $ P_1 $ y los valores de $ A $, $ B $, y $ C $ en la ecuación del plano:

$$ -2x + 5y + 3z + D = 0. $$

Sustituyendo $ x = 3 $, $ y = -2 $, y $ z = 5 $ obtenemos:

$$ -2(3) + 5(-2) + 3(5) + D = 0, $$
$$ -6 - 10 + 15 + D = 0, $$
$$ D = 1. $$

Por lo tanto, la ecuación del plano buscado es:

$$ -2x + 5y + 3z + 1 = 0. $$

Esta es la ecuación del plano que pasa por el punto $ (3, -2, 5) $ y que es perpendicular al vector $ (-2, 5, 3) $.

- ***Ejemplo 02.*** Hallar la ecuación del plano que pasa por el punto $P_1(-2,-1,5)$ y es perpendicular a la recta $l$ determinada por los puntos $P_2(2,-1,2)$ y $P_3(-3,1,-2)$.

Para encontrar la ecuación de un plano que pasa por un punto $ P_1(-2, -1, 5) $ y es perpendicular a una recta $ l $, necesitamos primero encontrar un vector normal al plano. Un vector normal al plano puede obtenerse a partir de la dirección de la recta $ l $, la cual a su vez se determina mediante un vector director formado por dos puntos dados en $ l $, en este caso $ P_2(2, -1, 2) $ y $ P_3(-3, 1, -2) $.

El vector director de la recta $ l $, que también será un vector normal al plano, es el vector que va de $ P_2 $ a $ P_3 $, que se puede calcular como:

$$ \vec{d} = P_3 - P_2 = (-3, 1, -2) - (2, -1, 2). $$

Realicemos este cálculo para obtener los componentes del vector normal al plano.

El vector normal al plano, obtenido de la dirección de la recta $ l $, es $ \vec{n} = (-5, 2, -4) $.

Ahora, con el punto $ P_1(-2, -1, 5) $ y el vector normal $ \vec{n} $, podemos escribir la ecuación del plano. Utilizando la fórmula general $ Ax + By + Cz + D = 0 $ y sustituyendo $ A $, $ B $, y $ C $ por los componentes del vector normal, obtenemos:

$$ -5x + 2y - 4z + D = 0. $$

Para encontrar $ D $, sustituimos el punto $ P_1 $ en la ecuación:

$$ -5(-2) + 2(-1) - 4(5) + D = 0, $$
$$ 10 - 2 - 20 + D = 0, $$
$$ D = 12. $$

Por lo tanto, la ecuación del plano es:

$$ -5x + 2y - 4z + 12 = 0. $$

Esta es la ecuación del plano que pasa por el punto $ P_1(-2, -1, 5) $ y es perpendicular a la recta $ l $ que pasa por los puntos $ P_2(2, -1, 2) $ y $ P_3(-3, 1, -2) $.

- ***Ejemplo 03.*** Hallar la ecuación del plano que pasa por los tres puntos no colineales $P_1(2,-1,1), P_2(-2,1,3)$ y $P_3(3,2,-2)$.

Para hallar la ecuación de un plano que pasa por tres puntos no colineales, primero necesitamos dos vectores que estén en el plano. Estos se pueden obtener restando las coordenadas de dos pares de puntos. Podemos usar $ P_1 $ y $ P_2 $ para el primer vector, y $ P_1 $ y $ P_3 $ para el segundo vector. 

Los vectores serán:

$$ \vec{v_1} = P_2 - P_1 = (-2 - 2, 1 - (-1), 3 - 1) $$
$$ \vec{v_2} = P_3 - P_1 = (3 - 2, 2 - (-1), -2 - 1) $$

Después de calcular estos vectores, podemos encontrar un vector normal al plano tomando el producto cruz de $ \vec{v_1} $ y $ \vec{v_2} $. El resultado será perpendicular a ambos vectores y, por lo tanto, normal al plano que los contiene.

Procedamos con los cálculos.

El vector normal al plano obtenido a partir del producto cruz de los vectores $ \vec{v_1} $ y $ \vec{v_2} $ es $ \vec{n} = [-12, -10, -14] $.

Ahora, para encontrar la ecuación del plano, utilizamos la fórmula general de la ecuación del plano $ Ax + By + Cz + D = 0 $, donde $ A $, $ B $, y $ C $ son las componentes del vector normal $ \vec{n} $. Así, tenemos:

$$ -12x - 10y - 14z + D = 0. $$

Para hallar $ D $, podemos sustituir las coordenadas de cualquiera de los tres puntos dados en la ecuación del plano. Usaremos el punto $ P_1(2, -1, 1) $:

$$ -12(2) - 10(-1) - 14(1) + D = 0, $$
$$ -24 + 10 - 14 + D = 0, $$
$$ D = 24 - 10 + 14. $$

Vamos a calcular $ D $ y así obtener la ecuación completa del plano.

El valor de $ D $ es 28. Por lo tanto, la ecuación del plano que pasa por los puntos $ P_1(2, -1, 1) $, $ P_2(-2, 1, 3) $, y $ P_3(3, 2, -2) $ es:

$$ -12x - 10y - 14z + 28 = 0. $$

Esta ecuación puede ser simplificada dividiendo todos los términos por -2 para obtener los coeficientes más pequeños posibles manteniendo la misma relación:

$$ 6x + 5y + 7z - 14 = 0. $$

## Trazas en un plano

Las trazas de un plano son las intersecciones de este con los planos coordenados, y nos permiten tener una visión más clara de la orientación del plano en el espacio tridimensional. Estas trazas se visualizan como rectas en los planos coordenados $ xy $, $ yz $, y $ zx $.

Considerando la ecuación general de un plano:

$$Ax + By + Cz + D = 0,$$

las trazas en cada plano coordenado se obtienen al establecer una de las variables $ x $, $ y $, o $ z $ en cero, de la siguiente manera:

- **Plano $ xy $ (cuando $ z = 0 $):** La traza en este plano se obtiene estableciendo $ z = 0 $ en la ecuación del plano. Esto nos da la ecuación de la recta en el plano $ xy $:

$$Ax + By + D = 0.$$

Esta es la ecuación de una recta en el plano $ xy $, que muestra la intersección del plano con el plano horizontal $ xy $.

- **Plano $ yz $ (cuando $ x = 0 $):** Análogamente, establecemos $ x = 0 $ para encontrar la traza en el plano $ yz $:

$$By + Cz + D = 0.$$

Esta es la ecuación de una recta en el plano $ yz $ y representa la proyección del plano sobre el plano vertical $ yz $.

- **Plano $ zx $ (cuando $ y = 0 $):** Finalmente, al hacer $ y = 0 $, encontramos la traza en el plano $ zx $:

$$Ax + Cz + D = 0.$$

Esta recta en el plano $ zx $ muestra cómo el plano dado corta el plano vertical $ zx $.

Las trazas nos permiten entender la inclinación y la orientación del plano en el espacio y son muy útiles, por ejemplo, para dibujar el plano en una representación tridimensional. También facilitan la comprensión de cómo un plano puede cortar a través de las diferentes dimensiones del espacio y ayudan en la solución de sistemas de ecuaciones que involucran múltiples planos.

## Intercepto de un plano con respecto a los ejes

Los interceptos de un plano con respecto a los ejes coordenados son los puntos donde el plano cruza los ejes $ x $, $ y $, y $ z $. Cada uno de estos puntos se encuentra al fijar las otras dos variables en cero y resolviendo para la variable restante. Estos puntos son útiles para esbozar el plano en un sistema de coordenadas y para entender su orientación espacial.

Dada la ecuación general de un plano:

$$ Ax + By + Cz + D = 0, $$

los interceptos se encuentran de la siguiente manera:

- **Intercepto en el eje $ x $ (cuando $ y = 0 $ y $ z = 0 $):** Al establecer $ y $ y $ z $ en cero, la ecuación se simplifica a $ Ax + D = 0 $. Resolviendo para $ x $, obtenemos:

$$ x = -\frac{D}{A}. $$

Si $ A $ es cero, el plano es paralelo al eje $ x $ y no hay intercepto con el eje $ x $.

- **Intercepto en el eje $ y $ (cuando $ x = 0 $ y $ z = 0 $):** De manera similar, si $ x $ y $ z $ son cero, tenemos $ By + D = 0 $. Resolviendo para $ y $, encontramos:

$$ y = -\frac{D}{B}. $$

Si $ B $ es cero, el plano es paralelo al eje $ y $ y no hay intercepto con el eje $ y $.

- **Intercepto en el eje $ z $ (cuando $ x = 0 $ y $ y = 0 $):** Por último, con $ x $ y $ y $ en cero, la ecuación se convierte en $ Cz + D = 0 $. Resolviendo para $ z $, se obtiene:

$$ z = -\frac{D}{C}. $$

Si $ C $ es cero, el plano es paralelo al eje $ z $ y no hay intercepto con el eje $ z $.

Los interceptos son comúnmente representados por los puntos $ (a, 0, 0) $, $ (0, b, 0) $, y $ (0, 0, c) $, donde $ a = -\frac{D}{A} $, $ b = -\frac{D}{B} $, y $ c = -\frac{D}{C} $. Estos valores representan las distancias desde el origen a los puntos donde el plano intercepta los ejes $ x $, $ y $, y $ z $, respectivamente. Si uno de los denominadores es cero, eso indica que el plano es paralelo al eje correspondiente y por lo tanto no hay intercepto real. Si $ D $ es cero, entonces uno o más de los interceptos estarán en el origen, indicando que el plano pasa por el origen del sistema de coordenadas.

## Forma simétrica del plano

La forma simétrica de la ecuación de un plano es una representación que relaciona cada coordenada del punto $ (x, y, z) $ con sus interceptos respectivos sobre los ejes $ x $, $ y $, y $ z $. Es útil para identificar rápidamente estos interceptos y para entender cómo varían las coordenadas del punto a lo largo del plano.

Partiendo de la ecuación general del plano:

$$
Ax + By + Cz + D = 0,
$$

y dividiendo cada término por $ -D $ (asumiendo que $ D \neq 0 $), obtenemos:

$$
\frac{Ax}{-D} + \frac{By}{-D} + \frac{Cz}{-D} - 1 = 0.
$$

Reescribiendo esta ecuación en términos de los interceptos $ a = -\frac{D}{A} $, $ b = -\frac{D}{B} $, y $ c = -\frac{D}{C} $, tenemos:

$$
\frac{x}{a} + \frac{y}{b} + \frac{z}{c} - 1 = 0,
$$

lo que simplificado da la forma simétrica de la ecuación del plano:

$$
\frac{x}{a} + \frac{y}{b} + \frac{z}{c} = 1.
$$

Esta forma simétrica nos dice que para cualquier punto $ (x, y, z) $ que esté en el plano, la suma de las fracciones de sus coordenadas divididas por sus interceptos correspondientes es igual a uno. Es importante destacar que esta forma solo es válida cuando $ A $, $ B $, y $ C $ no son cero, ya que de lo contrario los interceptos no estarían definidos (el plano sería paralelo a uno de los ejes y no habría intersección).

La forma simétrica de la ecuación del plano es particularmente útil cuando se trabaja con planos en problemas de geometría descriptiva y cuando se requiere una rápida visualización de las proporciones del plano con respecto a los ejes del sistema de coordenadas. Además, proporciona una manera elegante de confirmar si un punto dado está sobre el plano simplemente verificando si la suma de las coordenadas del punto, cada una dividida por su intercepto correspondiente, es igual a uno.

- ***Ejemplo 04:*** Para el siguiente plano encontrar los puntos de intersección con los ejes coordenados y las ecuaciones de sus trazas sobre los planos coordenados, además encuentre la forma simétrica de su ecuación $m_1: 4 x+6 y+3 z-12=0$.

Para resolver el ejemplo proporcionado, seguiremos estos pasos:

1. **Encontrar los puntos de intersección con los ejes coordenados**

El plano dado es:

$$4x + 6y + 3z - 12 = 0$$

- **Intersección con el eje $ x $ (cuando $ y = 0 $ y $ z = 0 $):**
$$4x - 12 = 0 \implies x = \frac{12}{4} = 3.$$
Punto de intersección: $ (3, 0, 0) $.

- **Intersección con el eje $ y $ (cuando $ x = 0 $ y $ z = 0 $):**
$$6y - 12 = 0 \implies y = \frac{12}{6} = 2.$$
Punto de intersección: $ (0, 2, 0) $.

- **Intersección con el eje $ z $ (cuando $ x = 0 $ y $ y = 0 $):**
$$3z - 12 = 0 \implies z = \frac{12}{3} = 4.$$
Punto de intersección: $ (0, 0, 4) $.


2. **Encontrar las ecuaciones de las trazas sobre los planos coordenados**

- **Plano $ xy $ (cuando $ z = 0 $):**
$$4x + 6y - 12 = 0$$

- **Plano $ yz $ (cuando $ x = 0 $):**
$$6y + 3z - 12 = 0$$

- **Plano $ zx $ (cuando $ y = 0 $):**
$$4x + 3z - 12 = 0$$


3. **Encontrar la forma simétrica de la ecuación del plano**

La forma simétrica se obtiene dividiendo cada término por $ -12 $:
$$
\frac{4x}{-12} + \frac{6y}{-12} + \frac{3z}{-12} - 1 = 0 \\
\frac{x}{-3} + \frac{y}{-2} + \frac{z}{-4} - 1 = 0 \\
\frac{x}{3} + \frac{y}{2} + \frac{z}{4} = 1.
$$

Así, la forma simétrica de la ecuación del plano $ m_1 $ es:
$$
\frac{x}{3} + \frac{y}{2} + \frac{z}{4} = 1.
$$

Estos resultados proporcionan una descripción completa del plano en términos de sus intersecciones con los ejes, las trazas en los planos coordenados y su representación en forma simétrica, facilitando su visualización y análisis en problemas geométricos y de diseño.

- ***Ejemplo 05.*** Hallar la ecuación del plano que pasa por el punto $P(2,1,-3)$ y es paralelo al plano $5 x-2 y+4 z-9=0$.

Para encontrar la ecuación de un plano que pasa por un punto dado y es paralelo a otro plano, podemos utilizar el hecho de que los planos paralelos comparten el mismo vector normal. Dado que el vector normal al plano dado $5x - 2y + 4z - 9 = 0$ es $[5, -2, 4]$, el plano que buscamos también tendrá este mismo vector normal.

***Dado:***
- Punto $P(2,1,-3)$
- Vector normal del plano paralelo: $[5, -2, 4]$

La ecuación general de un plano con un vector normal $n = (A, B, C)$ que pasa por un punto $(x_0, y_0, z_0)$ es:
$$ A(x - x_0) + B(y - y_0) + C(z - z_0) = 0. $$

Sustituyendo el punto $P(2, 1, -3)$ y el vector normal $[5, -2, 4]$, tenemos:
$$ 5(x - 2) - 2(y - 1) + 4(z + 3) = 0. $$

Expandamos y simplifiquemos esta ecuación:
$$ 5x - 10 - 2y + 2 + 4z + 12 = 0. $$
$$ 5x - 2y + 4z + 4 = 0. $$

La ecuación del plano que pasa por el punto $ (2, 1, -3) $ y es paralelo al plano $5x - 2y + 4z - 9 = 0$ es:
$$ 5x - 2y + 4z + 4 = 0. $$

Esta es la ecuación buscada, y describe un plano que comparte el mismo vector normal con el plano dado y pasa a través del punto especificado.

- ***Ejemplo 06.*** Hallar la ecuación del plano perpendicular al plano $X Y$ y que pasa por los puntos $P_1(1,5,-3)$ y $P_2(-5,-4,11)$.

Para encontrar la ecuación de un plano que es perpendicular al plano $XY$ y que pasa por dos puntos dados, debemos entender que un plano perpendicular al plano $XY$ es paralelo al eje $z$, y su ecuación solo involucra las variables $x$ y $y$, ya que la componente $z$ es irrelevante (no influye en la orientación del plano en relación con el eje $z$). La ecuación del plano, por tanto, no tendrá el término $z$, y se expresará como:

$$
Ax + By + D = 0
$$

Usamos los puntos $P_1$ y $P_2$ para determinar los coeficientes $A$ y $B$, sabiendo que ambos deben satisfacer la ecuación del plano.

**Formulamos un sistema de ecuaciones basado en estos puntos:**
1. Sustituyendo $P_1(1, 5)$:
   $$ A(1) + B(5) + D = 0 $$
   $$ A + 5B + D = 0 $$

2. Sustituyendo $P_2(-5, -4)$:
   $$ A(-5) + B(-4) + D = 0 $$
   $$ -5A - 4B + D = 0 $$

**Resolver este sistema de ecuaciones lineales:**

Para resolver este sistema, eliminamos $D$ sumando las ecuaciones:

$$ A + 5B + D + (-5A - 4B + D) = 0 $$
$$ -4A + B + 2D = 0 $$

Ahora, aislamos $D$ en una de las ecuaciones originales para eliminarlo en la otra. Por ejemplo, de la ecuación de $P_1$:

$$ D = -A - 5B $$

Sustituimos en la ecuación $ -4A + B + 2D = 0 $:

$$ -4A + B + 2(-A - 5B) = 0 $$
$$ -4A + B - 2A - 10B = 0 $$
$$ -6A - 9B = 0 $$
$$ 2A + 3B = 0 $$
$$ A = -\frac{3}{2}B $$

Ahora, sustituimos $A = -\frac{3}{2}B$ en $D = -A - 5B$:

$$ D = -\left(-\frac{3}{2}B\right) - 5B $$
$$ D = \frac{3}{2}B - 5B $$
$$ D = -\frac{7}{2}B $$

Finalmente, para encontrar valores específicos para $A$, $B$, y $D$, podemos establecer $B = 2$ (elección arbitraria para simplificar), entonces $A = -3$ y $D = -7$. Esto nos da la ecuación:

$$ -3x + 2y - 7 = 0 $$
$$ 3x - 2y + 7 = 0 $$

La ecuación del plano que es perpendicular al plano $XY$ y pasa por los puntos $P_1$ y $P_2$ es:
$$ 3x - 2y + 7 = 0 $$

- ***Ejemplo 07.*** Hallar la ecuación del plano que pasa por el punto $P(5,2,-3)$ y es perpendiculat a cada uno de los planos $2 x-y+2 z-9=0$ y $x+3 y-5 z+3=0$.

El plano que estamos buscando es perpendicular a dos planos dados, lo cual significa que su vector normal será perpendicular a los vectores normales de ambos planos. Utilizamos los determinantes para encontrar los componentes del vector normal al plano que buscamos.

***Vectores normales de los planos dados:***
- **Primer plano**: $2x - y + 2z - 9 = 0$ con vector normal $\vec{n}_1 = [2, -1, 2]$.
- **Segundo plano**: $x + 3y - 5z + 3 = 0$ con vector normal $\vec{n}_2 = [1, 3, -5]$.

***Uso de determinantes para encontrar el vector normal del plano buscado:***
Calculamos el producto cruz de $\vec{n}_1$ y $\vec{n}_2$ mediante determinantes:

1. **Componente $i$ ($x$-componente)**
   $$
   \left|\begin{array}{cc}
   -1 & 2 \\
   3 & -5
   \end{array}\right| = (-1)(-5) - (2)(3) = 5 - 6 = -1
   $$

2. **Componente $j$ ($y$-componente)**
   $$
   \left|\begin{array}{cc}
   2 & 2 \\

   \end{array}\right| = (2)(-5) - (2)(1) = -10 - 2 = -12
   $$

3. **Componente $k$ ($z$-componente)**
   $$
   \left|\begin{array}{cc}
   2 & -1 \\
   1 & 3
   \end{array}\right| = (2)(3) - (-1)(1) = 6 + 1 = 7
   $$

El vector normal resultante es $[1, -12, 7]$.

## Ángulo entre dos planos

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Plano02.PNG?raw=true" width="350" />
</p>

El ángulo entre dos planos en el espacio tridimensional puede ser entendido como el ángulo entre sus vectores normales. En la geometría analítica, este ángulo se puede calcular utilizando el producto escalar (también conocido como producto punto) de los vectores normales de los dos planos.

El producto escalar de dos vectores $\overrightarrow{a}$ y $\overrightarrow{b}$ se define como:

$$
\overrightarrow{a} \cdot \overrightarrow{b} = |\overrightarrow{a}| |\overrightarrow{b}| \cos(\theta)
$$

donde:
- $|\overrightarrow{a}|$ es la magnitud o longitud del vector $\overrightarrow{a}$,
- $|\overrightarrow{b}|$ es la magnitud o longitud del vector $\overrightarrow{b}$,
- $\theta$ es el ángulo entre los vectores $\overrightarrow{a}$ y $\overrightarrow{b}$.

Para calcular el ángulo $\theta$ entre dos vectores normales $\overrightarrow{n_1}$ y $\overrightarrow{n_2}$, reorganizamos la fórmula para resolver para $\cos(\theta)$:

$$
\cos(\theta) = \frac{\overrightarrow{n_1} \cdot \overrightarrow{n_2}}{|\overrightarrow{n_1}| |\overrightarrow{n_2}|}
$$

El producto escalar $\overrightarrow{n_1} \cdot \overrightarrow{n_2}$ se calcula como:

$$
\overrightarrow{n_1} \cdot \overrightarrow{n_2} = n_{1x}n_{2x} + n_{1y}n_{2y} + n_{1z}n_{2z}
$$

Y las magnitudes de los vectores normales se calculan como:

$$
|\overrightarrow{n_1}| = \sqrt{n_{1x}^2 + n_{1y}^2 + n_{1z}^2}
$$
$$
|\overrightarrow{n_2}| = \sqrt{n_{2x}^2 + n_{2y}^2 + n_{2z}^2}
$$

Una vez que tenemos $\cos(\theta)$, podemos encontrar el ángulo $\theta$ usando la función arccos (la función inversa del coseno):

$$
\theta = \arccos\left(\frac{\overrightarrow{n_1} \cdot \overrightarrow{n_2}}{|\overrightarrow{n_1}| |\overrightarrow{n_2}|}\right)
$$

Es importante notar que el ángulo resultante siempre será el menor de los dos ángulos posibles entre los planos y estará en el rango de $0^\circ$ a $90^\circ$. Si se desea el ángulo obtuso, se puede restar el ángulo calculado de $180^\circ$.

Además, cuando los planos son perpendiculares, el producto escalar será cero, y cuando son paralelos, el coseno será $\pm 1$, lo que indica un ángulo de $0^\circ$ o $180^\circ$ entre ellos.

- ***Ejemplo08:*** Encontrar el ángulo entre los planos $m_1: 3x + 2y + 8z - 9 = 0$ y $m_2: 5x + 6y - 2z + 3 = 0$

Para encontrar el ángulo entre los dos planos dados, utilizamos sus vectores normales. Los vectores normales pueden obtenerse directamente de los coeficientes de las variables $x$, $y$, y $z$ en las ecuaciones de los planos.

## Intersección entre dos planos

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Plano03.PNG?raw=true" width="250" />
</p>

La intersección de dos planos no paralelos en el espacio tridimensional es una recta. Cuando dos planos se cortan, sus ecuaciones pueden usarse para definir unívocamente la recta de intersección. Las ecuaciones de los planos en su forma general son:

$$
\begin{aligned}
& m_1: A_1 x + B_1 y + C_1 z + D_1 = 0 \\
& m_2: A_2 x + B_2 y + C_2 z + D_2 = 0
\end{aligned}
$$

Para que dos planos se intersecten, deben cumplir con las siguientes condiciones:

1. **No son paralelos**: Esto significa que sus vectores normales no son múltiplos escalares el uno del otro. Matemáticamente, esto se expresa como $ A_1/B_1 \neq A_2/B_2 $ o $ B_1/C_1 \neq B_2/C_2 $, lo que implica que el ángulo $ \theta $ entre los vectores normales $ \vec{n}_1 $ y $ \vec{n}_2 $ no es cero (o sea, $ \theta \neq 0 $).

2. **No son coincidentes**: Esto significa que no tienen exactamente las mismas ecuaciones o no son múltiplos escalares uno del otro.

Cuando se resuelve el sistema formado por las ecuaciones de los dos planos, el resultado es una serie de soluciones que representan los puntos a través de los cuales pasa la recta de intersección. Este sistema generalmente es compatible e indeterminado, ya que hay infinitos puntos a lo largo de la recta que satisfacen ambas ecuaciones.

Para expresar la recta de intersección de forma paramétrica o simétrica, se puede hacer lo siguiente:

- Seleccionar una variable para ser el parámetro, comúnmente $ t $, y expresar las otras dos en función de este parámetro.
- Resolver el sistema de ecuaciones de los dos planos para dos de las variables en términos de la tercera.
- Si es posible, eliminar el parámetro para obtener la ecuación simétrica de la recta.

La ecuación paramétrica de la recta tiene la forma:
$$
\begin{aligned}
& x = x_0 + at \\
& y = y_0 + bt \\
& z = z_0 + ct
\end{aligned}
$$
donde $ (x_0, y_0, z_0) $ es un punto por el que pasa la recta y $ (a, b, c) $ es la dirección de la recta, la cual se puede obtener también como el producto cruz de los vectores normales de los planos.

La ecuación simétrica se deriva de la paramétrica dividiendo cada una de las ecuaciones paramétricas por sus coeficientes direccionales correspondientes, asumiendo que son distintos de cero.

Conocer la recta de intersección entre dos planos es útil en diversas aplicaciones, como la construcción de objetos tridimensionales en diseño asistido por computadora, la solución de problemas en la geometría del espacio y en la física, especialmente en mecánica y óptica, donde las líneas de acción y las trayectorias suelen ser rectas que resultan de la intersección de planos de fuerzas o superficies reflectantes/refractantes.

- ***Ejemplo 09:*** Obtener la ecuación de la recta producto de la  intersección de los planos $𝑚1 : 𝑥 − 𝑦 − 𝑧 = 0$ y $𝑚2 : 𝑥 − 2𝑦 + 𝑧 − 1 = 0$.

Para encontrar la ecuación de la recta que es la intersección de los dos planos dados, $m_1: x - y - z = 0$ y $m_2: x - 2y + z - 1 = 0$, necesitamos resolver el sistema de ecuaciones que estos forman. En este caso, como estamos en tres dimensiones y solo tenemos dos ecuaciones, esperamos que la solución del sistema nos dé una recta.

Un método para resolver este sistema es expresar dos de las variables en términos de la tercera y luego usar esta tercera variable como un parámetro libre para las ecuaciones paramétricas de la recta. Veamos cómo hacerlo paso a paso.

***Paso 1: Reescribir las ecuaciones:*** Reescribimos las ecuaciones de los planos

$$ m_1: x - y - z = 0$$
$$ m_2: x - 2y + z - 1 = 0$$

***Paso 2: Resolver el sistema de ecuaciones*** Podemos sumar $ m_1 $ y $ m_2 $ para eliminar la variable $ z $. Sumando $ m_1 $ y $ m_2 $:
   $$ (x - y - z) + (x - 2y + z - 1) = 0 $$
   $$ 2x - 3y - 1 = 0 $$

Podemos resolver esta ecuación para $ x $ o $ y $ en términos del otro. Escojamos $ y $ como nuestro parámetro libre y expresemos $ x $ en términos de $ y $:

$$ 2x = 3y + 1 $$
$$ x = \frac{3}{2}y + \frac{1}{2} $$

Ahora, sustituimos $ x $ en $ m_1 $ para encontrar $ z $ en términos de $ y $:

$$ \frac{3}{2}y + \frac{1}{2} - y - z = 0 $$
$$ \frac{1}{2}y + \frac{1}{2} - z = 0 $$
$$ z = \frac{1}{2}y + \frac{1}{2} $$

***Paso 3: Ecuaciones paramétricas de la recta:*** Ahora que hemos expresado $ x $ y $ z $ en términos de $ y $, podemos escribir las ecuaciones paramétricas de la recta usando $ y $ como el parámetro (denotado como $ t $):

$$ x = \frac{3}{2}t + \frac{1}{2} $$
$$ y = t $$
$$ z = \frac{1}{2}t + \frac{1}{2} $$

***Resultado: Ecuación paramétrica de la recta de intersección:*** La ecuación paramétrica de la recta de intersección es entonces

$$
\begin{aligned}
x &= \frac{3}{2}t + \frac{1}{2} \\
y &= t \\
z &= \frac{1}{2}t + \frac{1}{2}
\end{aligned}
$$

Donde $ t $ es el parámetro que recorre la recta. Estas ecuaciones describen la serie de puntos por los que pasa la recta que es la intersección de los dos planos dados.

## Distancia de un punto a un plano

La distancia más corta de un punto a un plano es perpendicular al plano y se puede calcular utilizando el vector normal al plano y la posición del punto en cuestión. 

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Plano04.PNG?raw=true" width="350" />
</p>

Consideremos un plano en el espacio definido por la ecuación:

$$ m: A x + B y + C z + D = 0 $$

donde $ A $, $ B $, y $ C $ son los coeficientes que corresponden al vector normal del plano, $ \vec{n} $, y $ D $ es el término constante. El vector normal es una componente esencial, ya que dicta la orientación del plano en el espacio y es perpendicular a cualquier vector que yace en el plano.

Ahora, sea $ P(x_1, y_1, z_1) $ un punto fuera del plano. La distancia $ d $ entre el punto $ P $ y el plano $ m $ es el largo del segmento perpendicular desde $ P $ hasta $ m $. Este segmento representa la hipotenusa en un triángulo rectángulo imaginario cuyos catetos yacen en el plano y en la línea perpendicular que pasa por $ P $.


1. **Vector normal $ \vec{n} $**: Este vector se obtiene directamente de los coeficientes de la ecuación del plano y está representado como $ \vec{n} = [A, B, C] $.

2. **Vector unitario $ \hat{n} $**: El vector normal unitario se calcula normalizando el vector normal $ \vec{n} $, dividiéndolo entre su magnitud. Es decir:
   $$
   \hat{n} = \frac{\vec{n}}{|\vec{n}|}
   $$

3. **Magnitud del vector normal**: La magnitud $ |\vec{n}| $ es la raíz cuadrada de la suma de los cuadrados de los componentes del vector normal, que es:
   $$
   |\vec{n}| = \sqrt{A^2 + B^2 + C^2}
   $$

4. **Vector de posición $ \vec{P} $**: Es el vector desde un punto conocido $ x_0, y_0, z_0 $ en el plano (usualmente se toma como el punto más cercano en el plano al origen para simplificar los cálculos) hasta el punto $ P(x_1, y_1, z_1) $ cuya distancia queremos calcular, y está dado por:

$$\vec{P}=\left(x_1-x_0\right) \hat{i} + \left(y_1-y_0\right) \hat{j} + \left(z_1-x_z\right) \hat{k}$$

5. **Proyección de $ \vec{P} $ sobre $ \hat{n} $**: La distancia más corta entre el plano y el punto $ P $ es la longitud de la proyección del vector $ \vec{P} $ sobre el vector unitario normal $ \hat{n} $, y se calcula como:
   $$
   d = |\vec{P} \cdot \hat{n}|
   $$

6. **Ecuación de la distancia**: La ecuación de la distancia se simplifica al tomar en cuenta que el punto $ x_0, y_0, z_0 $ que se utiliza para definir el vector de posición $ \vec{P} $ es en realidad un punto del plano, por lo que satisface la ecuación del plano $ Ax_0 + By_0 + Cz_0 + D = 0 $. Por lo tanto, la ecuación final de la distancia $ d $ del punto $ P $ al plano $ m $ es:
   $$
   d = \left|\frac{A x_1 + B y_1 + C z_1 + D}{\sqrt{A^2 + B^2 + C^2}}\right|
   $$

Donde $ A, B, C $ son los coeficientes de la ecuación del plano y $ D $ es el término constante. La distancia $ d $ se obtiene tomando el valor absoluto de esta expresión, ya que la distancia es siempre un valor positivo. Este resultado se basa en la fórmula del producto punto y la proyección de un vector sobre otro, y es una aplicación directa del teorema de Pitágoras en el espacio tridimensional.

Distancia de un punto a un plano
Para un plano $m$ conocido $\left(\begin{array}{lll}\vec{n} & y & x_0\end{array}\right)$ y un punto $P\left(x_1, y_1, z_1\right)$ dado.
La distancia $d$ mas corta entre el plano y el punto $P$ es una distancia paralela al vector normal $\vec{n}$ al plano $m$.
Donde esta distancia $d$ se obtiene a partir de la proyección del vector $\vec{P}$ en $d=|\vec{P} \cdot \hat{n}|$ el vector unitario $\vec{n}$.
donde:
$$
\begin{gathered}
\hat{n}=\frac{\vec{n}}{|\vec{n}|} \Rightarrow \hat{n}=\frac{n_x \hat{\imath}+n_y \hat{\jmath}+n_z \hat{\mathrm{k}}}{\sqrt{n_x^2+n_y^2+n_z^2}} \\
\vec{P}=\left(x_1-x_0\right) \hat{\imath}+\left(y_1-y_0\right) \hat{\jmath}+\left(z_1-z_0\right) \hat{\mathrm{k}}, \text { sustituyendo } \\
d=\left|\frac{n_x\left(x_1-x_0\right)+n_y\left(y_1-y_0\right)+n_z\left(z_1-z_0\right)}{\sqrt{n_x^2+n_y^2+n_z^2}}\right| \\
d=\left|\frac{A x_1+B y_1+C z_1+D}{\sqrt{A^2+B^2+C^2}}\right|
\end{gathered}
$$

## Familia de Planos

Una familia de planos es un conjunto de planos que comparten una o más propiedades comunes, como un punto específico o una orientación particular. Las familias de planos tienen aplicaciones en múltiples campos, desde la geometría analítica hasta la ingeniería, la arquitectura y el diseño gráfico, donde la relación entre las superficies juega un papel crucial. Veamos algunas de las familias de planos más relevantes:

### Familia de Planos que Pasan por un Punto Conocido

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Plano05.PNG?raw=true" width="250" />
</p>

Una familia de planos que pasa por un punto específico comparte ese punto en común. El punto puede considerarse un vértice o un centro desde el cual "irradian" los diferentes planos.

Para un punto dado $P(x_1, y_1, z_1)$, la ecuación del plano que pasa por este punto puede escribirse de la siguiente manera:

$$A(x - x_1) + B(y - y_1) + C(z - z_1) = 0$$

Aquí, $A$, $B$, y $C$ representan los coeficientes del vector normal al plano, $\vec{n} = (A, B, C)$, que dicta la orientación del plano. La ecuación también se puede expandir para obtener la forma estándar:

$$Ax + By + Cz = Ax_1 + By_1 + Cz_1$$

El punto de intersección, $P(x_1, y_1, z_1)$, se denomina punto o vértice de radiación, ya que es el punto donde todas las superficies de la familia convergen. Cambiando los valores de $A$, $B$, y $C$, se pueden generar infinitos planos que pasan por $P$ con diferentes inclinaciones.

### Familia de Planos Paralelos

Otra familia común es la de los planos paralelos, que mantienen una distancia constante entre sí. Tienen el mismo vector normal pero difieren en el término constante, lo que significa que sus ecuaciones son similares salvo por el valor de $ D $. La ecuación de un plano general puede expresarse como:

$$ Ax + By + Cz + D = 0 $$

Donde el vector normal, $ \vec{n} = (A, B, C) $, es el mismo para cada plano de la familia. Solo el valor de $ D $ cambia, lo que permite generar múltiples planos con la misma orientación pero diferentes posiciones.


por ejemplo, los planos dados por las ecuaciones: 
$$m_1:7x+2y+3z+10=0$$ 
y 
$$m_2:7x+2y+3z-5=0$$

son paralelos.

### Familia de Planos Perpendiculares a una Recta

Los planos perpendiculares a una recta determinada forman otra familia interesante. Si una recta dada tiene dirección $ \vec{d} = (a, b, c) $, cualquier plano cuyo vector normal sea paralelo o antiparalelo a $ \vec{d} $ será perpendicular a la recta.

La ecuación del plano perpendicular a la recta puede escribirse en la forma:

$$ a(x - x_0) + b(y - y_0) + c(z - z_0) = 0 $$

donde $ (x_0, y_0, z_0) $ es un punto en el plano.

### Familia de Planos que Contienen la Recta Generada por la Intersección de dos Planos

Cuando dos planos en el espacio tridimensional se intersectan, lo hacen en una línea recta. Esta línea representa el conjunto de puntos que satisfacen simultáneamente las ecuaciones de ambos planos. La familia de planos que pasan por esta línea de intersección pueden describirse como una combinación lineal de las ecuaciones de los dos planos originales.

Dado el conjunto de ecuaciones:

$$ m_1: A_1 x + B_1 y + C_1 z + D_1 = 0 $$
$$ m_2: A_2 x + B_2 y + C_2 z + D_2 = 0 $$

la ecuación de cualquier plano que contiene la línea de intersección de los planos $ m_1 $ y $ m_2 $ se expresa como:

$$ A_1 x + B_1 y + C_1 z + D_1 + k(A_2 x + B_2 y + C_2 z + D_2) = 0 $$

donde $ k $ es un parámetro arbitrario que ajusta el peso de las contribuciones de cada plano. Al expandir esta ecuación, se observa que todos los planos de esta familia comparten la recta de intersección, ya que satisfacen ambas ecuaciones simultáneamente.

***Propiedades***

1. **Estructura General:** La ecuación representa la combinación de las ecuaciones de ambos planos, y el parámetro $ k $ proporciona la flexibilidad para generar infinitos planos que pasan por la misma línea de intersección.

2. **Vector Normal:** El vector normal de cualquiera de estos planos es una combinación lineal de los vectores normales de los dos planos originales.

3. **Aplicación:** Esta familia de planos puede utilizarse para identificar superficies que comparten un borde común, diseñar estructuras espaciales que pasan por una misma línea de referencia, o crear modelos que requieran superficies alineadas.

**Propiedad en común:** Estos dos planos son idénticos.

***b)***

**Propiedad en común:** Estos planos son paralelos entre sí.

**Propiedad en común:** Estos planos comparten un punto en común: $ (2, 7, -2) $.

Para encontrar la ecuación del plano $ m_o $ que pasa por la intersección de los planos $ m_1 $ y $ m_2 $ y tiene una distancia de $\sqrt{3}$ con respecto al origen, seguiremos estos pasos:

***Paso 1: Escribir la ecuación del plano $ m_o $ -*** La ecuación del plano $ m_o $ que pasa por la intersección de los planos $ m_1 $ y $ m_2 $ puede representarse como una combinación lineal de las ecuaciones de ambos planos:
$$ m_o: m_1 + k m_2 = x + y + z + 3 + k (2x + 3y - 5z) = 0 $$
Expandamos esta expresión para obtener la forma estándar:
$$ x + y + z + 3 + 2kx + 3ky - 5kz = 0 $$
$$ (1 + 2k)x + (1 + 3k)y + (1 - 5k)z + 3 = 0 $$

Esta es la ecuación general del plano $ m_o $. Necesitamos encontrar el valor de $ k $ que haga que el plano esté a una distancia de $\sqrt{3}$ del origen.

***Paso 2: Calcular el vector normal de $ m_o $ -*** El vector normal del plano $ m_o $ es:
$$ \vec{n} = (1 + 2k, 1 + 3k, 1 - 5k) $$

***Paso 3: Usar la fórmula de distancia -*** La distancia entre el plano $ m_o $ y el origen es dada por la fórmula:

$$ d = \frac{|A \cdot 0 + B \cdot 0 + C \cdot 0 + D|}{\sqrt{A^2 + B^2 + C^2}} = \frac{|3|}{\sqrt{(1 + 2k)^2 + (1 + 3k)^2 + (1 - 5k)^2}} $$

Queremos que la distancia sea igual a $\sqrt{3}$:
$$ \sqrt{3} = \frac{|3|}{\sqrt{(1 + 2k)^2 + (1 + 3k)^2 + (1 - 5k)^2}} $$

***Paso 4: Resolver para $ k $ -*** Despejemos el valor de $ k $ que satisface esta ecuación y obtendremos la ecuación final del plano $ m_o $.

1. **Expresión del vector normal:**

2. **Magnitud del vector normal:**

   $$

3. **Ecuación de distancia:**

   $$
