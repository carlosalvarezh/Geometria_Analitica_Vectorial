### Magnitudes escalares y vectoriales

Las magnitudes que emplearemos en este curso ser√°n de dos tipos: [escalares](https://en.wikipedia.org/wiki/Scalar_(mathematics)) y [vectoriales](https://en.wikipedia.org/wiki/Euclidean_vector).

Una *magnitud escalar* es aquella que queda completamente determinada con un n√∫mero y sus correspondientes unidades, y una *magnitud vectorial* es aquella que, adem√°s de un valor num√©rico y sus unidades (m√≥dulo) debemos especificar su direcci√≥n y sentido.

La elecci√≥n de un escalar o un vector para representar una magnitud f√≠sica depende de la naturaleza de la misma; si estamos describiendo la temperatura de una habitaci√≥n, la densidad de un cuerpo, su masa... necesitaremos representarlas mediante un n√∫mero. Por el contrario, cuando trabajemos con magnitudes como la fuerza, la velocidad, la aceleraci√≥n, el campo el√©ctrico, etc., emplearemos vectores.

Un vector en el plano (bidimensional o 2D) o  el espacio (tridimensional o 3D) est√° caracterizado por dos/tres n√∫meros que se denominan componentes o coordenadas del vector.

Las componentes de un vector ser√°n en general diferentes dependiendo del sistema de coordenadas que utilicemos para expresarlas, pero siempre es posible relacionarlas de una manera sistem√°tica.

### Definici√≥n de Vector

Un **vector** es una entidad matem√°tica que se define por dos caracter√≠sticas principales: **magnitud** (m√≥dulo o longitud) y **direcci√≥n**. Se representa gr√°ficamente como un segmento de recta orientado, que tiene un punto de inicio (o cola) y un punto final (o cabeza).

Matem√°ticamente, un vector en el espacio tridimensional se representa a menudo como una terna ordenada de n√∫meros $\vec{a} = (x, y, z)$ o, de manera equivalente, como una combinaci√≥n lineal de los vectores unitarios en las direcciones $X$, $Y$ y $Z$, $ \vec{a} = x\hat{i} + y\hat{j} + z\hat{k} $, donde $\hat{i}$, $\hat{j}$ y $\hat{k}$ son los vectores unitarios en las direcciones $X$, $Y$ y $Z$, respectivamente.

En este curso emplearemos el sistema de coordenadas cartesianas. Como se indic√≥ anteriormente, el sistema de coordenadas cartesianas est√° constituido por dos ejes (tres si trabajamos en tres dimensiones) perpendiculares entre s√≠ que se cortan en un punto llamado origen.

Las componentes cartesianas de un vector son las proyecciones de dicho vector sobre cada uno de los ejes. 

$$\vec{a} = (a_x, a_y, a_z)$$

La magnitud (o longitud) de un vector $\vec{a} = (a_x, a_y, a_z)$ se calcula con la norma euclidiana: 

$$\| \vec{a} \| = \sqrt{a_x^2 + a_y^2 + a_z^2}$$

Representa la "distancia" del vector desde el origen hasta su punto terminal (observe que se cumple el teorema de Pit√°goras).

### Vector unitario

Un [**vector unitario**](https://en.wikipedia.org/wiki/Unit_vector) es un vector que tiene una magnitud (o longitud) de exactamente $1$ unidad, manteniendo una direcci√≥n espec√≠fica. Estos vectores son particularmente √∫tiles en matem√°ticas y f√≠sica porque proporcionan una manera de representar puramente la direcci√≥n de un vector, sin preocuparse por su magnitud.

#### Caracter√≠sticas de un Vector Unitario

- ***Magnitud:*** La magnitud de un vector unitario es siempre $1$. Esto significa que su longitud en un gr√°fico es siempre de una unidad est√°ndar, independientemente de su direcci√≥n.

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Vector06Unitario.gif?raw=true" width="200" />
</p>
<p>&nbsp;</p>

2. ***Direcci√≥n:*** La direcci√≥n de un vector unitario es su caracter√≠stica m√°s importante. Indica la orientaci√≥n del vector en el espacio, pero no su tama√±o o longitud. La direcci√≥n se mantiene igual que la del vector original del cual se deriva el vector unitario.
<p>&nbsp;</p>

3. ***Normalizaci√≥n:*** Para convertir un vector cualquiera en un vector unitario, se realiza un proceso llamado normalizaci√≥n. La normalizaci√≥n implica dividir cada componente del vector por la magnitud del vector original. Matem√°ticamente, si $\vec{a} = (x, y, z)$ es un vector, su vector unitario $\hat{a}$ se encuentra dividiendo $\vec{a}$ por su magnitud: $\hat{a} = \frac{\vec{a}}{\| \vec{a} \|}$.
<p>&nbsp;</p>

4. ***Uso de Vectores Unitarios Est√°ndar:*** En el sistema de coordenadas cartesianas $3D$, los vectores unitarios est√°ndar son $\hat{i}$, $\hat{j}$ y $\hat{k}$, que representan las direcciones de los ejes $X$, $Y$ y $Z$, respectivamente. Cualquier vector en el plano puede expresarse como una combinaci√≥n lineal de $\hat{i}$, $\hat{j}$ y $\hat{k}$.
<p>&nbsp;</p>

5. **Aplicaciones:**
   - Los vectores unitarios se utilizan para indicar direcciones en f√≠sica, como la direcci√≥n de una fuerza, velocidad o aceleraci√≥n.
   - En ingenier√≠a y gr√°ficos por computadora, se utilizan para representar orientaciones y direcciones de movimiento o fuerza.

#### Producto escalar

El producto escalar de dos vectores $u$ y $v$ que forman un √°ngulo $\phi$ se define como:

***Caso $2D$:***

Observe que el producto escalar de dos vectores no es un vector, es un n√∫mero (un escalar). Adem√°s el producto escalar de dos vectores perpendiculares es nulo. Se deducen entonces los siguientes resultados:

\begin{aligned}
& \vec{i} \cdot \vec{i}=\vec{j} \cdot \vec{j} = 1 \\
& \vec{i} \cdot \vec{j}=\vec{j} \cdot \vec{i} = 0
\end{aligned}

Si los vectores est√°n expresados en componentes, en dos dimensiones y aplicando los resultados anteriores se obtiene que:

$$\vec{u} \cdot \vec{v}=\left(u_x \vec{i}+u_y \vec{j}\right) \cdot\left(v_x \vec{i}+v_y \vec{j}\right)=u_x v_x+u_y v_y$$

El producto escalar de dos vectores posee la propiedad [conmutativa](https://en.wikipedia.org/wiki/Commutative_property).

***Caso $3D$:***

El producto escalar de dos vectores $\vec{u}$ y $\vec{v}$ que forman un √°ngulo $\phi$ en tres dimensiones se define de manera similar a como se hace en dos dimensiones. Sin embargo, en 3D se considera tambi√©n la componente a lo largo del eje $Z$.

Al igual que en el caso bidimensional, el producto escalar en tres dimensiones resulta en un n√∫mero (un escalar), no en otro vector. Tambi√©n se mantiene la propiedad de que el producto escalar de dos vectores perpendiculares es cero. Esto lleva a los siguientes resultados:

\begin{aligned}
& \vec{i} \cdot \vec{i}=\vec{j} \cdot \vec{j}=\vec{k} \cdot \vec{k} = 1 \\
& \vec{i} \cdot \vec{j}=\vec{j} \cdot \vec{k}=\vec{k} \cdot \vec{i} = 0
\end{aligned}

En tres dimensiones, si los vectores $\vec{u}$ y $\vec{v}$ est√°n expresados en componentes como $\vec{u} = u_x \vec{i} + u_y \vec{j} + u_z \vec{k}$ y $\vec{v} = v_x \vec{i} + v_y \vec{j} + v_z \vec{k}$, entonces el producto escalar se obtiene como:

$$\vec{u} \cdot \vec{v} = \left(u_x \vec{i} + u_y \vec{j} + u_z \vec{k}\right) \cdot \left(v_x \vec{i} + v_y \vec{j} + v_z \vec{k}\right) = u_x v_x + u_y v_y + u_z v_z$$


**Propiedad del √Ångulo en el Producto Escalar**

El producto escalar entre dos vectores no solo proporciona informaci√≥n sobre su magnitud, sino que tambi√©n puede usarse para determinar el √°ngulo $\phi$ entre ellos. Esta propiedad es v√°lida tanto en dos como en tres dimensiones.

Si se tienen dos vectores $\vec{u}$ y $\vec{v}$, el producto escalar entre ellos se define como:

$$\vec{u} \cdot \vec{v} = |\vec{u}| |\vec{v}| \cos(\phi)$$

Donde $|\vec{u}|$ y $|\vec{v}|$ son las magnitudes (o normas) de los vectores $\vec{u}$ y $\vec{v}$ respectivamente, y $\phi$ es el √°ngulo entre los dos vectores. Esta ecuaci√≥n se puede reorganizar para hallar el √°ngulo $\phi$:

$$\phi = \cos^{-1}\left(\frac{\vec{u} \cdot \vec{v}}{|\vec{u}| |\vec{v}|}\right)$$

Este c√°lculo del √°ngulo es especialmente √∫til en diversas aplicaciones como:

1. **Determinar la Ortogonalidad**: Si el producto escalar es $0$, entonces $\phi = 90^{\circ}$, lo que indica que los vectores son ortogonales (perpendiculares entre s√≠).

2. **An√°lisis de Direcciones**: Puede usarse para verificar si dos vectores est√°n en la misma direcci√≥n (√°ngulo de $0^{\circ}$) o en direcciones opuestas (√°ngulo de $180^{\circ}$).

***Nota:*** al producto escalar entre dos vectores tambi√©n se le conoce como producto punto.

##### Ejemplos producto escalar

***Ejemplo 1.*** obtener el √°ngulo ùúÉ entre las l√≠neas $ùëô_1 = 3ùë• + 4ùë¶ ‚àí 8$ y $ùëô_2 = 4ùë• ‚àí 3ùë¶ + 9 = 0$

Vamos a resolver este ejemplo de varias formas diferentes:

- ***Tangente del √Ångulo entre las rectas (Usando Pendientes)***

En el primer m√©todo, calculamos el √°ngulo usando la f√≥rmula de la tangente del √°ngulo entre dos l√≠neas, que se basa en sus pendientes:

$$ \tan(\theta) = \left| \frac{m_1 - m_2}{1 + m_1 m_2} \right| = \left| \frac{\frac{-3}{4} - \frac{4}{3}}{1 + \frac{-3}{4}\frac{4}{3}} \right| =  \left| \frac{\frac{-25}{12}}{1 - 1} \right|= \left| \frac{-25}{0} \right|$$

Al denominador ser cero, tenemos una indeterminaci√≥n, es decir $\tan(\theta) \rightarrow \infty$. Esto quiere decir que se tiene una as√≠ntota vertical, por lo que se concluye que ambas rectas son perpendiculares: $l_1 \perp l_2$

- ***Perpendicularidad entre rectas (concepto)***

Del enunciado se observa que las pendientes de las rectas $l_1$ y $l_2$, son respectivamente, $m_1=-3/4$ y $m_2=4/3$, y recordando del cap√≠tulo de *la recta* que si se tiene qu√©  $m_1 \times m_2 = -1$, entonces ambas rectas son perpediculares entre s√≠. Reemplazando los valores de las pendientes $m_1 \times m_2 = -3/4 \times 4/3 = -1$. Se comprueba, nuevamente, que $l_1 \perp l_2$.


- ***Producto Escalar (Usando Vectores Direccionales)***

Por √∫ltimo, usamos el producto escalar de los vectores direccionales de las l√≠neas:

$$ \cos(\theta) = \frac{\vec{v_1} \cdot \vec{v_2}}{|\vec{v_1}| |\vec{v_2}|} $$

Con $\vec{v_1} = (3, 4)$ y $\vec{v_2} = (4, -3)$, calculemos el producto escalar entre ellos:

$$\vec{v_1} \cdot \vec{v_2} = 3 \times 4 + 4 \times (-3) = 12 - 12 = 0$$

Observemos que el producto punto es cero, por lo que quiere decir que ambas rectas son perpendiculares entre s√≠, confirmando los resultados previamente encontrados por los otros m√©todos.

#### Producto de un vector por un escalar

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Vector11.PNG?raw=true" width="150" />
</p>

En tres dimensiones, un vector $\vec{v}$ se puede expresar en t√©rminos de sus componentes a lo largo de los ejes $x$, $y$, $z$, es decir, $\vec{v} = v_x \vec{i} + v_y \vec{j} + v_z \vec{k}$. El producto de este vector por un escalar $\alpha$ resulta en un nuevo vector $\vec{w}$ cuyas componentes son escaladas por $\alpha$. Matem√°ticamente, se expresa como:

$$\alpha \vec{v} = \alpha (v_x \vec{i} + v_y \vec{j} + v_z \vec{k}) = (\alpha v_x) \vec{i} + (\alpha v_y) \vec{j} + (\alpha v_z) \vec{k} = w_x \vec{i} + w_y \vec{j} + w_z \vec{k}$$

Donde $\vec{w} = \alpha \vec{v}$ es el vector resultante de la multiplicaci√≥n.

Este proceso se puede visualizar como un estiramiento o una compresi√≥n del vector original en todas las direcciones, manteniendo su direcci√≥n pero cambiando su magnitud. Si el escalar es mayor que $1$, el vector se alarga (estira); si est√° entre $0$ y $1$, se acorta (comprime); y si es negativo, adem√°s de cambiar su longitud, el vector cambia de direcci√≥n (se invierte).

Es importante destacar que:

1. **Mantenimiento de la Direcci√≥n**: La direcci√≥n del vector original se mantiene, a menos que el escalar sea negativo, en cuyo caso el vector se invierte.
   
2. **Cambio en la Magnitud**: La magnitud del vector cambia en funci√≥n del valor absoluto del escalar.

3. **Caso Especial de Cero**: Si el escalar es cero, el vector resultante es el vector nulo, independientemente del vector original.

En aplicaciones pr√°cticas, el producto de un vector por un escalar es √∫til en f√≠sica (por ejemplo, para describir el escalado de velocidades o fuerzas).

#### Producto vectorial

El producto vectorial de dos vectores que forman un √°ngulo $\phi$ es otro vector, de direcci√≥n perpendicular al plano formado por los dos vectores, sentido el que da la regla de la mano derecha y m√≥dulo el que se especifica a continuaci√≥n:

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Vector10.gif?raw=true" width="150" />
</p>

$$w = uv\sin(\phi)$$

El producto vectorial **no posee la propiedad conmutativa**, ya que se cumple que:

$$\vec{u} \times \vec{v} = -\vec{v} \times \vec{u}$$

Adem√°s, se cumple que el producto vectorial de dos vectores paralelos es nulo. Se obtienen entonces las siguientes relaciones ($2D$):

\begin{aligned}
& \vec{i} \times \vec{i}=\vec{j} \times \vec{j}=\vec{k} \times \vec{k}=0 \\
& \vec{i} \times \vec{j}=\vec{k} \quad \vec{j} \times \vec{k}=\vec{i} \quad \vec{k} \times i=\vec{j}
\end{aligned}

Si los vectores vienen expresados en componentes el producto vectorial se calcula desarrollando el determinante:

$$
\vec{u} \times \vec{v}=\left|\begin{array}{ccc}
\vec{i} & \vec{j} & \vec{k} \\
u_x & u_y & u_z \\
v_x & v_y & v_z
\end{array}\right|
$$

**Producto Vectorial en Tres Dimensiones**

En tres dimensiones, el producto vectorial de dos vectores $\vec{u}$ y $\vec{v}$, que forman un √°ngulo $\phi$, es un vector $\vec{w}$ que tiene las siguientes caracter√≠sticas:

1. **Direcci√≥n**: Perpendicular al plano formado por $\vec{u}$ y $\vec{v}$.
2. **Sentido**: Determinado por la regla de la mano derecha: si los dedos de la mano derecha apuntan de $\vec{u}$ a $\vec{v}$, el pulgar apunta en la direcci√≥n de $\vec{w}$.
3. **M√≥dulo**: El m√≥dulo de $\vec{w}$ es igual a $uv\sin(\phi)$, donde $u$ y $v$ son los m√≥dulos de $\vec{u}$ y $\vec{v}$ respectivamente.

El producto vectorial en $3D$, al igual que en otras dimensiones, **no es conmutativo**. Se cumple que:

$$\vec{u} \times \vec{v} = -\vec{v} \times \vec{u}$$

Esto significa que el cambio en el orden de los vectores invierte la direcci√≥n del vector resultante.

Otra propiedad importante es que el producto vectorial de dos vectores paralelos es un vector nulo. En el caso de los vectores unitarios en el espacio tridimensional, se tienen las siguientes relaciones:

\begin{aligned}
& \vec{i} \times \vec{i}=\vec{j} \times \vec{j}=\vec{k} \times \vec{k}=0 \\
& \vec{i} \times \vec{j}=\vec{k}, \quad \vec{j} \times \vec{k}=\vec{i}, \quad \vec{k} \times \vec{i}=\vec{j} \\
& \vec{j} \times \vec{i}=-\vec{k}, \quad \vec{k} \times \vec{j}=-\vec{i}, \quad \vec{i} \times \vec{k}=-\vec{j}
\end{aligned}

Para calcular el producto vectorial cuando los vectores est√°n expresados en componentes, se utiliza el determinante de una matriz $3\times3$ formada por los vectores unitarios $\vec{i}, \vec{j}, \vec{k}$ en la primera fila, y las componentes de $\vec{u}$ y $\vec{v}$ en la segunda y tercera fila respectivamente:

$$
\vec{u} \times \vec{v}=\left|\begin{array}{ccc}
\vec{i} & \vec{j} & \vec{k} \\
u_x & u_y & u_z \\
v_x & v_y & v_z
\end{array}\right|
$$

El resultado de este determinante es el vector $\vec{w}$, que es perpendicular a ambos $\vec{u}$ y $\vec{v}$ y cuyo m√≥dulo es el √°rea del paralelogramo que forman estos dos vectores.


Otras propiedades del producto vectorial es:

#### Triple producto vectorial

El triple producto vectorial es una operaci√≥n vectorial que se aplica sobre tres vectores en el espacio tridimensional. Esta operaci√≥n se define como el producto vectorial entre un vector y el producto vectorial de otros dos vectores.

$$ \vec{u} \times (\vec{v} \times \vec{w}) $$

donde $\vec{u}$, $\vec{v}$, y $\vec{w}$ son vectores en el espacio tridimensional.

***Interpretaci√≥n Geom√©trica***

El triple producto vectorial $\vec{u} \times (\vec{v} \times \vec{w})$ produce un vector que es perpendicular al vector $\vec{u}$ y que est√° en el plano formado por $\vec{v}$ y $\vec{w}$.

***Propiedad Matem√°tica***

Una propiedad importante del triple producto vectorial es que puede reescribirse utilizando una combinaci√≥n lineal de los otros dos vectores. Esto se conoce como la *f√≥rmula del desarrollo del triple producto vectorial* y se expresa como:

$$ \vec{u} \times (\vec{v} \times \vec{w}) = (\vec{u} \cdot \vec{c})\vec{v} - (\vec{u} \cdot \vec{v})\vec{w} $$

Esta f√≥rmula muestra que el resultado del triple producto vectorial es una combinaci√≥n de los vectores $\vec{v}$ y $\vec{w}$, ponderada por los productos escalares de $\vec{u}$ con $\vec{w}$ y $\vec{v}$, respectivamente.

***Procedimiento de c√°lculo***

Consideremos los vectores $\vec{u} = (u_1, u_2, u_3)$, $\vec{v} = (v_1, v_2, v_3)$, y $\vec{w} = (w_1, w_2, w_3)$. El triple producto vectorial se calcula como sigue:

1. Primero calculamos el producto vectorial de $\vec{v} \times \vec{w}$.
2. Luego, el resultado se usa para calcular el producto vectorial con $\vec{u}$.

Este proceso puede requerir el uso de la determinante para el c√°lculo del producto vectorial y luego la aplicaci√≥n de la f√≥rmula del desarrollo del triple producto vectorial.

***Aplicaciones***

El triple producto vectorial es √∫til en f√≠sica y ingenier√≠a, especialmente en situaciones que implican rotaciones y momentos angulares. Por ejemplo, en la din√°mica de cuerpos r√≠gidos, puede usarse para encontrar direcciones perpendiculares a ciertos planos o para calcular componentes de fuerzas en direcciones espec√≠ficas.

***Aplicaciones Pr√°cticas***

El m√©todo del paralelogramo se utiliza en f√≠sica para la suma de diferentes tipos de vectores como fuerzas, velocidades y desplazamientos. Permite a los ingenieros y cient√≠ficos visualizar c√≥mo dos fuerzas o movimientos interact√∫an y resultan en un efecto combinado.

#### Ley de cosenos

Cuando sumamos dos vectores $\vec{u}$ y $\vec{v}$ utilizando el m√©todo del paralelogramo, a menudo nos interesa no solo la direcci√≥n del vector resultante, sino tambi√©n su magnitud. $$|\vec{w}|=\sqrt{|\vec{u}|^2 + |\vec{v}|^2 - 2|\vec{u}||\vec{v}|\cos(\theta)}$$


***Interpretaci√≥n Geom√©trica***

En el contexto del paralelogramo, $\theta$ es el √°ngulo menor entre los vectores $\vec{u}$ y $\vec{v}$. La f√≥rmula toma en cuenta este √°ngulo para calcular correctamente la magnitud de $\vec{w}$. Si los vectores son perpendiculares (√°ngulo de 90 grados), el t√©rmino $-2|\vec{u}||\vec{v}|\cos(\theta)$ se anula, y la f√≥rmula se reduce al teorema de Pit√°goras.

***Ejemplo Pr√°ctico***

Si tenemos dos vectores, $\vec{u}$ y $\vec{v}$, con magnitudes de 5 y 3 unidades, respectivamente, y forman un √°ngulo de 60¬∞, entonces la magnitud de $\vec{w}$ ser√≠a:

$$|\vec{w}|=\sqrt{5^2 + 3^2 - 2 \times 5 \times 3 \times \cos(60^\circ)}$$

Este c√°lculo nos da la longitud del vector resultante.


#### Descomposicion de un vector

$$ \theta = \arccos\left(\frac{\vec{A} \cdot \vec{B}}{|\vec{A}| |\vec{B}|}\right) = \arccos\left(\frac{13}{\sqrt{86234}}\right) \approx 1.5265 \text{ rads}$$

Por √∫ltimo, pasamos ese valor de radianes a grados:

## Ecuaciones de la Recta en 3D

### Ecuaciones de la recta en 3D

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/Recta3D.PNG?raw=true" width="200" />
</p>

Consideremos la recta $ l $, que pasa por dos puntos dados $ P_0 $ y $ P $. Estos puntos se definen por sus vectores posici√≥n $ \overrightarrow{r_0} $ y $ \vec{r} $ respectivamente, de la siguiente manera:
$$
\begin{aligned}
& \overrightarrow{r_0} = x_0 \hat{\mathbf{i}} + y_0 \hat{\mathbf{j}} + z_0 \hat{\mathbf{k}}, \\
& \vec{r} = x \hat{\mathbf{i}} + y \hat{\mathbf{j}} + z \hat{\mathbf{k}}.
\end{aligned}
$$

Introducimos un vector $ \vec{A} $, que representa el desplazamiento desde $ P_0 $ a $ P $, definido como:
$$
\vec{A} = \vec{r} - \overrightarrow{r_0}.
$$

El vector direcci√≥n $ \vec{V} $, paralelo a la recta $ l $, con componentes conocidas $ a $, $ b $ y $ c $, se expresa as√≠:
$$
\vec{V} = a \hat{\mathbf{i}} + b \hat{\mathbf{j}} + c \hat{\mathbf{k}}.
$$

Establecemos que $ \vec{A} $ es proporcional a $ \vec{V} $ a trav√©s de un escalar $ t $:
$$
\vec{A} = \vec{r} - \overrightarrow{r_0} = t \vec{V}.
$$

Despejando $ \vec{r} $, obtenemos la ecuaci√≥n vectorial de la recta:
$$
\vec{r} = \overrightarrow{r_0} + t \vec{V}.
$$

Desarrollando todos los componentes vectoriales, obtenemos la ***ecuaci√≥n vectorial de la recta**** en 3D:
$$
x \hat{\mathbf{i}} + y \hat{\mathbf{j}} + z \hat{\mathbf{k}} = x_0 \hat{\mathbf{i}} + y_0 \hat{\mathbf{j}} + z_0 \hat{\mathbf{k}} + t(a \hat{\mathbf{i}} + b \hat{\mathbf{j}} + c \hat{\mathbf{k}}).
$$

Esta ecuaci√≥n vectorial puede convertirse en su forma escalar, resultando en tres ***[ecuaciones param√©tricas](https://en.wikipedia.org/wiki/Parametric_equation)***:
$$
\begin{aligned}
& x = x_0 + at, \\
& y = y_0 + bt, \\
& z = z_0 + ct.
\end{aligned}
$$

Despejando $ t $ en cada una de las ecuaciones param√©tricas y luego igualando estos valores, obtenemos la ***ecuaci√≥n de la recta en su forma sim√©trica***:
$$
\frac{x - x_0}{a} = \frac{y - y_0}{b} = \frac{z - z_0}{c}.
$$

Esta ecuaci√≥n sim√©trica proporciona una descripci√≥n completa y concisa de la recta en un espacio tridimensional.

***Ejemplo01:*** Encontrar las tres formas de la recta que pasa por los puntos $A(‚àí5,4,1)$ y $B(9,10,‚àí5)$.

Para encontrar las tres formas de la ecuaci√≥n de la recta que pasa por los puntos $ A(-5, 4, 1) $ y $ B(9, 10, -5) $, primero obtenemos el vector direcci√≥n:

- **Calcular el vector direcci√≥n $ \vec{V} $**: Este vector se obtiene restando las coordenadas de $ A $ de las de $ B $.  

$$ \vec{V} = B - A = (9 - (-5), 10 - 4, -5 - 1) = (14, 6, -6)$$

- **Ecuaci√≥n Vectorial**: Utilizando el punto $ A $ y el vector $ \vec{V} $, la ecuaci√≥n vectorial es de la forma $ \vec{r} = \vec{r_0} + t\vec{V} $, donde $ \vec{r_0} $ es el vector posici√≥n de $ A $, llegando a:

$$\vec{r} = (-5 \hat{\mathbf{i}} + 4 \hat{\mathbf{j}} + 1 \hat{\mathbf{k}}) + t(14 \hat{\mathbf{i}} + 6 \hat{\mathbf{j}} - 6 \hat{\mathbf{k}})$$

- **Ecuaci√≥n Param√©trica**: Desarrollamos la ecuaci√≥n vectorial en sus componentes $ x $, $ y $, y $ z $.
   $$
   \begin{aligned}
   x &= -5 + 14t, \\
   y &= 4 + 6t, \\
   z &= 1 - 6t.
   \end{aligned}
   $$

- **Ecuaci√≥n Sim√©trica**: Despejamos $ t $ en las ecuaciones param√©tricas y las igualamos, la ecuaci√≥n sim√©trica es:
   $$\frac{x + 5}{14} = \frac{y - 4}{6} = \frac{z - 1}{-6}$$

Estas son las tres formas de la ecuaci√≥n de la recta que pasa por los puntos $ A(-5, 4, 1) $ y $ B(9, 10, -5) $.

### Distancia de un punto a una l√≠nea recta en 3D

Para calcular la distancia $ D $ de un punto $ P(x, y, z) $ a una l√≠nea recta $ l_1 $ en el espacio tridimensional, el proceso involucra varios pasos geom√©tricos y vectoriales.

<p float="center">
  <img src="https://github.com/carlosalvarezh/Geometria_Analitica_Vectorial/blob/main/Images/DistanciaPuntoRecta3D.PNG?raw=true" width="350" />
</p>


Primero, seleccionamos un punto $ Q $ sobre la recta $ l_1 $ y definimos un vector direcci√≥n $ \vec{V_1} $ que caracteriza la recta. A continuaci√≥n, trazamos un vector $ \overrightarrow{QP} $ que conecta el punto $ Q $ con el punto $ P $. La distancia $ D $ que buscamos es la longitud del segmento perpendicular que va desde $ P $ hasta la recta $ l_1 $.

La clave para determinar $ D $ es usar la propiedad del producto vectorial. El m√≥dulo del producto vectorial de dos vectores es igual al producto de sus m√≥dulos por el seno del √°ngulo $ \theta $ entre ellos: 
$$ |\vec{d} \times \overrightarrow{QP}| = |\vec{d}| \cdot |\overrightarrow{QP}| \cdot \sin \theta. $$

El √°ngulo $ \theta $ aqu√≠ es el formado entre el vector direcci√≥n $ \vec{d} $ y $ \overrightarrow{QP} $. Para hallar la distancia $ D $, necesitamos expresar $ \sin \theta $ en t√©rminos de cantidades conocidas. Recordando que $ \sin \theta $ es la relaci√≥n entre el lado opuesto a $ \theta $ en un tri√°ngulo rect√°ngulo y su hipotenusa, y en nuestro caso, el lado opuesto es la distancia $ D $ y la hipotenusa es $ |\overrightarrow{QP}| $, tenemos:
$$ \sin \theta = \frac{D}{|\overrightarrow{QP}|}. $$

Al sustituir esta relaci√≥n en nuestra ecuaci√≥n inicial y despejar $ D $, obtenemos la f√≥rmula para calcular la distancia del punto $ P $ a la recta:
$$ D = \frac{|\vec{d} \times \overrightarrow{QP}|}{|\vec{d}|}. $$

Esta f√≥rmula permite calcular la distancia m√°s corta desde un punto dado hasta una l√≠nea recta en el espacio tridimensional.

***Ejemplo 02:*** Encontrar la distancia entre el punto $(3,-2,1)$ y la recta $l_1: \frac{-x+2}{-2}=\frac{3 y+3}{-6}=z$

Para encontrar la distancia entre un punto y una recta en 3D, seguimos los pasos mencionados anteriormente. En este caso, necesitamos calcular la distancia entre el punto $ P(3, -2, 1) $ y la recta $ l_1 $ definida por la ecuaci√≥n

1. **Encontrar un punto en la recta $ l_1 $**: Podemos obtener un punto $ Q $ sobre la recta $ l_1 $ eligiendo un valor para el par√°metro (digamos, $ t = 0 $) y sustituy√©ndolo en la ecuaci√≥n de la recta.

2. **Determinar el vector direcci√≥n $ \vec{d} $ de la recta**: La ecuaci√≥n de la recta en su forma sim√©trica nos da el vector direcci√≥n.

3. **Calcular el vector $ \overrightarrow{QP} $**: Usamos las coordenadas de $ P $ y $ Q $ para encontrar este vector.

4. **Aplicar la f√≥rmula de la distancia**: La distancia $ D $ se encuentra usando la f√≥rmula 
   $$ D = \frac{|\vec{d} \times \overrightarrow{QP}|}{|\vec{d}|}. $$

Comencemos calculando un punto $ Q $ sobre la recta $ l_1 $ y el vector direcci√≥n $ \vec{d} $.

Ahora, aplicaremos la f√≥rmula de la distancia:
$$ D = \frac{|\vec{d} \times \overrightarrow{QP}|}{|\vec{d}|}. $$

Para ello, primero calcularemos el producto vectorial $ \vec{d} \times \overrightarrow{QP} $ y luego encontraremos su magnitud y la magnitud de $ \vec{d} $. Con estos valores, podremos calcular la distancia $ D $. Procedamos con estos c√°lculos.

### Distancia entre dos rectas en 3D

La medici√≥n de la distancia m√°s corta $D$ entre dos rectas, $l_1$ y $l_2$, en un espacio tridimensional es una tarea que involucra conceptos de geometr√≠a vectorial. Esta distancia se define como la longitud del segmento de l√≠nea m√°s corto que conecta las dos rectas, siendo perpendicular a ambas.

Para calcular $D$, usamos la siguiente f√≥rmula:

$$
D = \left| \frac{\left( \vec{V_1} \times \vec{V_2} \right) \cdot \vec{P_1P_2}}{|\vec{V_1} \times \vec{V_2}|} \right|,
$$

donde:
- $\vec{V_1}$ es el vector direcci√≥n de la recta $l_1$.
- $\vec{V_2}$ es el vector direcci√≥n de la recta $l_2$.
- $\vec{P_1P_2}$ es el vector que conecta un punto conocido $P_1$ en la recta $l_1$ con un punto conocido $P_2$ en la recta $l_2$.

Es crucial que el producto vectorial $\vec{V_1} \times \vec{V_2}$ sea no nulo (es decir, $|\vec{V_1} \times \vec{V_2}| \neq 0$), lo cual asegura que las rectas no son paralelas.

La f√≥rmula proporciona la distancia proyectando el vector que une los dos puntos en la direcci√≥n normal al plano definido por las dos rectas. Esta proyecci√≥n nos da la longitud del segmento m√°s corto entre $l_1$ y $l_2$, que es perpendicular a ambas rectas. Este enfoque es particularmente √∫til en problemas de geometr√≠a tridimensional donde las distancias directas no son f√°cilmente visualizables o medibles.

***Ejemplo 03:*** Encontrar la distancia entre las rectas $l_1: \frac{x-4}{2}=\frac{y-5}{4}=\frac{z+4}{3}$ y $l_2: \frac{x+1}{3}=\frac{y-7}{2}=\frac{z-5}{6}$

Para encontrar la distancia entre las rectas $ l_1 $ y $ l_2 $ usaremos la f√≥rmula para la distancia entre dos rectas en el espacio tridimensional:

$$D = \left| \frac{\left( \vec{V_1} \times \vec{V_2} \right) \cdot \vec{P_1P_2}}{|\vec{V_1} \times \vec{V_2}|} \right|$$

donde:
- $ \vec{V_1} $ y $ \vec{V_2} $ son los vectores direcci√≥n de las rectas $ l_1 $ y $ l_2 $, respectivamente.
- $ \vec{P_1P_2} $ es un vector que conecta un punto conocido $ P_1 $ en la recta $ l_1 $ con un punto conocido $ P_2 $ en la recta $ l_2 $.

Primero, identificaremos los vectores direcci√≥n y los puntos en las rectas. Despu√©s, aplicaremos la f√≥rmula. Empezaremos extrayendo los vectores direcci√≥n y los puntos de las ecuaciones de las rectas.

Los vectores direcci√≥n y los puntos en las rectas son:

- Para $ l_2 $: Vector direcci√≥n $ \vec{V_2} = (3, 2, 6) $ y punto $ P_2 = (-1, 7, 5) $.
- El vector $ \vec{P_1P_2} $ que conecta $ P_1 $ y $ P_2 $ es $ (-5, 2, 9) $.

Ahora, procederemos a calcular el producto vectorial $ \vec{V_1} \times \vec{V_2} $, su magnitud y finalmente aplicaremos la f√≥rmula de la distancia.


$$
\vec{V_1} \times \vec{V_2} = \begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
2 & 4 & 3 \\
3 & 2 & 6
\end{vmatrix}
$$

El determinante se calcula expandiendo a lo largo de la primera fila:

$$
\begin{align*}
\vec{V_1} \times \vec{V_2} &= \hat{i} \cdot \begin{vmatrix}
4 & 3 \\
2 & 6
\end{vmatrix} - \hat{j} \cdot \begin{vmatrix}
2 & 3 \\
3 & 6
\end{vmatrix} + \hat{k} \cdot \begin{vmatrix}
2 & 4 \\
3 & 2
\end{vmatrix} \\
&= \hat{i} \cdot (4 \cdot 6 - 3 \cdot 2) - \hat{j} \cdot (2 \cdot 6 - 3 \cdot 3) + \hat{k} \cdot (2 \cdot 2 - 4 \cdot 3) \\
&= \hat{i} \cdot (24 - 6) - \hat{j} \cdot (12 - 9) + \hat{k} \cdot (4 - 12) \\
&= \hat{i} \cdot 18 - \hat{j} \cdot 3 + \hat{k} \cdot (-8) \\
&= (18, -3, -8)
\end{align*}
$$

y su magnitud es:

$$|\vec{V_1} \times \vec{V_2}| = \sqrt{18^2 + (-3)^2 + (-8)^2} = \sqrt{324 + 9 + 64} = \sqrt{397} \approx 19.925$$

Ahora realizaremos el c√°lculo del producto escalar

$$ \left(\vec{V_1} \times \vec{V_2} \right) \cdot \vec{P_1P_2} = 18 \times -5 + (-3) \times 2 + (-8) \times 9 = -90 + (-6) + (-72) = -168$$


Reemplazando estos valores en la f√≥rmula para el c√°lculo de la distancia entre dos rectas en el espacio, se tiene:

$$D = \left| \frac{-168}{19.925} \right| \approx 8.43 \text{ unidades}$$

Con lo que la distancia m√°s corta entre las rectas $l_1: \frac{x-4}{2}=\frac{y-5}{4}=\frac{z+4}{3}$ y $l_2: \frac{x+1}{3}=\frac{y-7}{2}=\frac{z-5}{6}$ es aproximadamente $ 8.43 $ unidades.
